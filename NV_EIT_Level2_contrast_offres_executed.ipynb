# NV-EIT: Level 2 contrast model with an off-resonant excited state

この notebook は、NV-EIT において隣接励起準位 \(|e_2\rangle\) が存在する場合の **Level 2** モデルを計算するためのものです。

Level 1 では、off-resonant coupling の効果を

$$
\gamma_{\mathrm{eff}}
=
\gamma_{g,\mathrm{eff}}
+
\frac{|\Omega_{c2}|^2}{4}
\frac{\gamma_{\mathrm{off}}}
{\Delta_{\mathrm{off}}^2+\gamma_{\mathrm{off}}^2}
$$

として、ground-state coherence の有効緩和に吸収していました。

Level 2 では、off-resonant coupling を単なる $\gamma_{\mathrm{eff}}$ の増大としてではなく、

$$
\Delta_g^{\mathrm{eff}}(\delta,\Delta_p)
=
\delta+i\gamma_g
-
\frac{|\Omega_{c2}|^2}{4\Delta_{e_2g_2}}
$$

として明示的に入れます。これにより、off-resonant coupling による **light shift** と **induced decoherence** を同時に扱えます。


## 1. 最小 4 準位モデル

考える準位は次の 4 つです。

$$
|g_2\rangle \leftrightarrow |e_1\rangle
\quad \mathrm{probe}
$$

$$
|g_1\rangle \leftrightarrow |e_1\rangle
\quad \mathrm{resonant\ control}
$$

$$
|g_1\rangle \leftrightarrow |e_2\rangle
\quad \mathrm{off-resonant\ control}
$$

隣接励起準位のエネルギー差を

$$
\omega_{e_2}-\omega_{e_1}
=
\Delta_{\mathrm{off}}
$$

とします。

weak probe limit では、EIT に対応する optical coherence を

$$
\sigma_{e_1g_2}^{(1)}
=
-
\frac{\rho_{g_2g_2}^{(0)}\Omega_{p1}}{2}
\frac{
\Delta_g^{\mathrm{eff}}
}{
\Delta_{e_1g_2}\Delta_g^{\mathrm{eff}}
-
|\Omega_{c1}|^2/4
}
$$

と置きます。

ここで

$$
\Delta_{e_1g_2}
=
\Delta_p+i\gamma_{e_1g_2}
$$

$$
\Delta_{e_2g_2}
=
\Delta_p-\Delta_{\mathrm{off}}+i\gamma_{\mathrm{off}}
$$

$$
\Delta_g^{\mathrm{eff}}
=
\delta+i\gamma_g
-
\frac{|\Omega_{c2}|^2}{4\Delta_{e_2g_2}}
$$

です。


## 2. light shift と induced decoherence

二光子共鳴付近で $\Delta_p \approx 0$ とすると、

$$
\Delta_{e_2g_2}
\simeq
-\Delta_{\mathrm{off}}+i\gamma_{\mathrm{off}}
$$

なので、

$$
-
\frac{|\Omega_{c2}|^2}{4\Delta_{e_2g_2}}
=
\frac{|\Omega_{c2}|^2}{4}
\frac{
\Delta_{\mathrm{off}}+i\gamma_{\mathrm{off}}
}{
\Delta_{\mathrm{off}}^2+\gamma_{\mathrm{off}}^2
}
$$

となります。

したがって

$$
\Delta_g^{\mathrm{eff}}
=
\delta
+
\delta_{\mathrm{LS}}
+
i\gamma_{\mathrm{eff}}
$$

ただし、

$$
\delta_{\mathrm{LS}}
=
\frac{|\Omega_{c2}|^2}{4}
\frac{
\Delta_{\mathrm{off}}
}{
\Delta_{\mathrm{off}}^2+\gamma_{\mathrm{off}}^2
}
$$

$$
\gamma_{\mathrm{eff}}
=
\gamma_g
+
\frac{|\Omega_{c2}|^2}{4}
\frac{
\gamma_{\mathrm{off}}
}{
\Delta_{\mathrm{off}}^2+\gamma_{\mathrm{off}}^2
}
$$

です。

したがって、以前の Level 1 の式は、Level 2 において $\Delta_g^{\mathrm{eff}}$ の虚部だけを取り出した近似に対応します。


## 3. 吸収と EIT contrast の定義

吸収を

$$
A(\delta)
\propto
\mathrm{Im}\left[\chi(\delta)\right]
$$

で定義します。

Level 2 では

$$
\chi(\delta)
\propto
-
\frac{
\Delta_g^{\mathrm{eff}}(\delta)
}{
\Delta_{e_1g_2}\Delta_g^{\mathrm{eff}}(\delta)
-
|\Omega_{c1}|^2/4
}
$$

を用います。

背景吸収は control field を切った場合の吸収として

$$
\chi_{\mathrm{bg}}
=
-\frac{1}{\Delta_{e_1g_2}}
$$

$$
A_{\mathrm{bg}}
=
\mathrm{Im}\left[\chi_{\mathrm{bg}}\right]
$$

とします。

EIT contrast は、light shift によって dip の中心が $ \delta=0 $ からずれる可能性があるため、

$$
A_{\mathrm{min}}
=
\min_{\delta} A(\delta)
$$

を用いて

$$
C_{\mathrm{EIT}}
=
1-
\frac{
A_{\mathrm{min}}
}{
A_{\mathrm{bg}}
}
$$

と定義します。

明示的には

$$
C_{\mathrm{EIT}}(\Omega_{c1})
=
1
-
\frac{
\min_{\delta}
\mathrm{Im}
\left[
-
\dfrac{
\Delta_g^{\mathrm{eff}}(\delta)
}{
\Delta_{e_1g_2}\Delta_g^{\mathrm{eff}}(\delta)
-
|\Omega_{c1}|^2/4
}
\right]
}{
\mathrm{Im}
\left[
-\dfrac{1}{\Delta_{e_1g_2}}
\right]
}
$$

です。


## 4. $ \Omega_{c2} $ のモデル

同じ control laser が resonant transition と off-resonant transition の両方に結合するなら、最も自然な仮定は

$$
\Omega_{c2}
=
\eta\Omega_{c1}
$$

です。

ここで

$$
\eta
=
\frac{
d_{\mathrm{off}}
}{
d_{\mathrm{res}}
}
\times
\mathrm{polarization/mixing\ factor}
$$

です。

ただし NV center では、$\eta$ は strain、磁場、偏光、励起状態混合、温度依存の phonon process に依存し得ます。そのため、この notebook では以下の 2 種類を比較します。

### constant-$\eta$ model

$$
\Omega_{c2}
=
\eta_0\Omega_{c1}
$$

### phenomenological growth model

$$
\Omega_{c2}
=
\eta_0\Omega_{c1}
\left[
1
+
a
\frac{
(\Omega_{c1}/\Delta_{\mathrm{off}})^m
}{
1+(\Omega_{c1}/\Delta_{\mathrm{off}})^m
}
\right]
$$

後者は、$\Omega_{c1}\sim\Delta_{\mathrm{off}}$ 付近で off-resonant な mixing が強まる可能性を表す現象論的モデルです。


```python
import numpy as np
import matplotlib.pyplot as plt

# ============================================================
# Units
# ============================================================
# This notebook uses MHz as the unit for all frequency-like quantities.
# If you want angular frequencies, use all parameters consistently in rad/s.
```

```python
# ============================================================
# Parameters
# ============================================================

params = {
    # optical coherence decay rate for |e1> <-> |g2>
    "gamma_e": 15.0,          # MHz

    # ground-state coherence decay rate
    "gamma_g": 0.05,          # MHz

    # detuning between neighboring excited states
    "Delta_off": 500.0,       # MHz

    # linewidth / dephasing of the off-resonant excited state
    "gamma_off": 100.0,       # MHz

    # one-photon detuning of the probe from |g2> -> |e1>
    "Delta_p": 0.0,           # MHz
}

# Omega_c2 = eta0 * Omega_c1 for the constant-eta model
eta0 = 1.0

# Parameters for phenomenological growth model
growth_a = 4.0
growth_m = 6

# scan range for two-photon detuning delta
delta_grid = np.linspace(-800.0, 800.0, 8001)  # MHz

# scan range for resonant control Rabi frequency Omega_c1
Omega_c1_grid = np.linspace(1.0, 1500.0, 400)  # MHz
```

```python
def omega_c2_constant(Omega_c1, eta0=1.0):
    """Constant transition-dipole-ratio model: Omega_c2 = eta0 * Omega_c1."""
    return eta0 * Omega_c1


def omega_c2_growth(Omega_c1, Delta_off=500.0, eta0=1.0, a=4.0, m=6):
    """
    Phenomenological growth model.

    Omega_c2 = eta0 * Omega_c1 * [1 + a x^m / (1 + x^m)]
    where x = Omega_c1 / Delta_off.
    """
    x = np.asarray(Omega_c1) / Delta_off
    return eta0 * np.asarray(Omega_c1) * (1.0 + a * x**m / (1.0 + x**m))


def delta_e1g2(Delta_p, gamma_e):
    """Complex detuning for |g2> -> |e1>."""
    return Delta_p + 1j * gamma_e


def delta_e2g2(Delta_p, Delta_off, gamma_off):
    """Complex detuning for |g2> -> |e2>."""
    return Delta_p - Delta_off + 1j * gamma_off


def delta_g_eff(delta, Omega_c2, Delta_p, Delta_off, gamma_off, gamma_g):
    """
    Effective ground-state detuning including the off-resonant excited state.

    Delta_g_eff = delta + i gamma_g - |Omega_c2|^2 / (4 Delta_e2g2)
    """
    De2 = delta_e2g2(Delta_p, Delta_off, gamma_off)
    return delta + 1j * gamma_g - (np.abs(Omega_c2)**2) / (4.0 * De2)


def chi_level2(delta, Omega_c1, Omega_c2, gamma_e, gamma_g, Delta_off, gamma_off, Delta_p=0.0):
    """
    Level 2 susceptibility up to an irrelevant proportionality constant.

    chi = - Delta_g_eff / (Delta_e1g2 * Delta_g_eff - |Omega_c1|^2/4)
    """
    De1 = delta_e1g2(Delta_p, gamma_e)
    Dg = delta_g_eff(delta, Omega_c2, Delta_p, Delta_off, gamma_off, gamma_g)
    return -Dg / (De1 * Dg - (np.abs(Omega_c1)**2) / 4.0)


def absorption_level2(delta, Omega_c1, Omega_c2, params):
    """Absorption proportional to Im[chi]."""
    chi = chi_level2(
        delta=delta,
        Omega_c1=Omega_c1,
        Omega_c2=Omega_c2,
        gamma_e=params["gamma_e"],
        gamma_g=params["gamma_g"],
        Delta_off=params["Delta_off"],
        gamma_off=params["gamma_off"],
        Delta_p=params["Delta_p"],
    )
    return np.imag(chi)


def background_absorption(params):
    """
    Background absorption with the control field off:
    chi_bg = -1 / Delta_e1g2.
    """
    De1 = delta_e1g2(params["Delta_p"], params["gamma_e"])
    return np.imag(-1.0 / De1)


def level1_contrast(Omega_c1, Omega_c2, params):
    """
    Level 1 contrast formula using only induced decoherence.
    This is included for comparison.
    """
    gamma_eff = (
        params["gamma_g"]
        + (np.abs(Omega_c2)**2 / 4.0)
        * params["gamma_off"]
        / (params["Delta_off"]**2 + params["gamma_off"]**2)
    )
    return (np.abs(Omega_c1)**2 / 4.0) / (
        params["gamma_e"] * gamma_eff + np.abs(Omega_c1)**2 / 4.0
    )


def level2_contrast(Omega_c1, Omega_c2, delta_grid, params):
    """
    Level 2 contrast obtained from the minimum of the absorption spectrum.
    """
    A = absorption_level2(delta_grid, Omega_c1, Omega_c2, params)
    A_bg = background_absorption(params)

    idx_min = np.argmin(A)
    A_min = A[idx_min]
    delta_min = delta_grid[idx_min]
    C = 1.0 - A_min / A_bg
    return C, A_min, A_bg, delta_min
```

## 5. 吸収スペクトルの例

以下では、$ \Omega_{c1} $ を増やしたときに、吸収スペクトル

$$
A(\delta)
=
\mathrm{Im}[\chi(\delta)]
$$

がどのように変化するかを見ます。

特に、Level 2 では $ \delta_{\mathrm{LS}} $ によって dip の位置がずれるため、$ \delta=0 $ だけを見るのではなく、

$$
A_{\mathrm{min}}
=
\min_{\delta} A(\delta)
$$

を計算します。


```python
# ============================================================
# Plot absorption spectra for several Omega_c1 values
# ============================================================

Omega_examples = [50.0, 200.0, 500.0, 800.0, 1200.0]  # MHz

plt.figure(figsize=(8, 5))

for Oc1 in Omega_examples:
    Oc2 = omega_c2_constant(Oc1, eta0=eta0)
    A = absorption_level2(delta_grid, Oc1, Oc2, params)
    plt.plot(delta_grid, A, label=fr"$\Omega_{{c1}}={Oc1:.0f}$ MHz")

plt.axvline(0.0, linestyle="--", linewidth=1)
plt.xlabel(r"Two-photon detuning $\delta$ (MHz)")
plt.ylabel(r"Absorption $A(\delta) \propto \mathrm{Im}[\chi]$")
plt.title("Level 2 absorption spectra: constant-eta model")
plt.legend()
plt.tight_layout()
plt.show()
```

## 6. contrast の $ \Omega_{c1}$ 依存性

Level 2 contrast は

$$
C_{\mathrm{EIT}}(\Omega_{c1})
=
1
-
\frac{
A_{\mathrm{min}}(\Omega_{c1})
}{
A_{\mathrm{bg}}
}
$$

で計算します。

ここで

$$
A_{\mathrm{min}}(\Omega_{c1})
=
\min_{\delta} A(\delta;\Omega_{c1})
$$

です。

比較のため、Level 1 の近似式

$$
C_{\mathrm{EIT}}^{\mathrm{Level\ 1}}
=
\frac{|\Omega_{c1}|^2/4}
{
\gamma_{e_1g_2}\gamma_{\mathrm{eff}}
+
|\Omega_{c1}|^2/4
}
$$

も同時に描画します。


```python
# ============================================================
# Compute contrast curves
# ============================================================

C_level1_const = []
C_level2_const = []
dip_position_const = []

C_level1_growth = []
C_level2_growth = []
dip_position_growth = []

for Oc1 in Omega_c1_grid:
    # constant eta
    Oc2_const = omega_c2_constant(Oc1, eta0=eta0)
    C1 = level1_contrast(Oc1, Oc2_const, params)
    C2, A_min, A_bg, delta_min = level2_contrast(Oc1, Oc2_const, delta_grid, params)
    C_level1_const.append(C1)
    C_level2_const.append(C2)
    dip_position_const.append(delta_min)

    # phenomenological growth
    Oc2_g = omega_c2_growth(
        Oc1,
        Delta_off=params["Delta_off"],
        eta0=eta0,
        a=growth_a,
        m=growth_m,
    )
    C1g = level1_contrast(Oc1, Oc2_g, params)
    C2g, A_min_g, A_bg_g, delta_min_g = level2_contrast(Oc1, Oc2_g, delta_grid, params)
    C_level1_growth.append(C1g)
    C_level2_growth.append(C2g)
    dip_position_growth.append(delta_min_g)

C_level1_const = np.asarray(C_level1_const)
C_level2_const = np.asarray(C_level2_const)
dip_position_const = np.asarray(dip_position_const)

C_level1_growth = np.asarray(C_level1_growth)
C_level2_growth = np.asarray(C_level2_growth)
dip_position_growth = np.asarray(dip_position_growth)
```

```python
# ============================================================
# Plot contrast vs Omega_c1 / Delta_off
# ============================================================

x = Omega_c1_grid / params["Delta_off"]

plt.figure(figsize=(8, 5))
plt.plot(x, C_level1_const, label="Level 1, constant eta")
plt.plot(x, C_level2_const, linestyle="--", label="Level 2, constant eta")
plt.plot(x, C_level1_growth, label="Level 1, eta growth")
plt.plot(x, C_level2_growth, linestyle="--", label="Level 2, eta growth")
plt.axvline(1.0, linestyle=":", linewidth=1, label=r"$\Omega_{c1}=\Delta_{\mathrm{off}}$")
plt.xlabel(r"$\Omega_{c1}/\Delta_{\mathrm{off}}$")
plt.ylabel(r"EIT contrast $C_{\mathrm{EIT}}$")
plt.ylim(-0.05, 1.05)
plt.title("EIT contrast from absorption minimum")
plt.legend()
plt.tight_layout()
plt.show()
```

## 7. light-shift による dip 位置の移動

Level 2 では、off-resonant coupling によって

$$
\delta_{\mathrm{LS}}
=
\frac{|\Omega_{c2}|^2}{4}
\frac{
\Delta_{\mathrm{off}}
}{
\Delta_{\mathrm{off}}^2+\gamma_{\mathrm{off}}^2
}
$$

が生じます。

そのため、EIT dip は一般に

$$
\delta = 0
$$

ではなく、おおよそ

$$
\delta
\simeq
-\delta_{\mathrm{LS}}
$$

付近に移動します。

以下では、数値的に得られた dip 位置

$$
\delta_{\mathrm{min}}
=
\arg\min_{\delta} A(\delta)
$$

をプロットします。


```python
# ============================================================
# Plot dip position vs Omega_c1 / Delta_off
# ============================================================

plt.figure(figsize=(8, 5))
plt.plot(x, dip_position_const, label="Level 2, constant eta")
plt.plot(x, dip_position_growth, label="Level 2, eta growth")
plt.axhline(0.0, linestyle="--", linewidth=1)
plt.axvline(1.0, linestyle=":", linewidth=1, label=r"$\Omega_{c1}=\Delta_{\mathrm{off}}$")
plt.xlabel(r"$\Omega_{c1}/\Delta_{\mathrm{off}}$")
plt.ylabel(r"Dip position $\delta_{\min}$ (MHz)")
plt.title("Light-shift-induced displacement of the EIT dip")
plt.legend()
plt.tight_layout()
plt.show()
```

## 8. Level 2 モデルの解釈

constant-$\eta$ model では、

$$
\Omega_{c2}
=
\eta_0\Omega_{c1}
$$

であり、off-resonant induced decoherence は

$$
\gamma_{\mathrm{off}}^{\mathrm{ind}}
=
\frac{\eta_0^2|\Omega_{c1}|^2}{4}
\frac{
\gamma_{\mathrm{off}}
}{
\Delta_{\mathrm{off}}^2+\gamma_{\mathrm{off}}^2
}
$$

です。

この場合、$ \gamma_{\mathrm{off}}^{\mathrm{ind}} $ は $ |\Omega_{c1}|^2 $ に比例しますが、EIT を開く項

$$
\frac{|\Omega_{c1}|^2}{4}
$$

も同じく $ |\Omega_{c1}|^2 $ に比例します。

そのため、$ \eta_0 $ と $\gamma_{\mathrm{off}} $ が一定である限り、contrast は多くの場合、単純に低下するというより、ある値に飽和します。

一方、実験的に  $ \Omega_{c1}\sim\Delta_{\mathrm{off}} $ 付近から contrast が低下する場合は、以下の追加要因が考えられます。

$$
\eta
=
\eta(\Omega_{c1},T,B,\varepsilon,\mathrm{polarization})
$$

が高 power 側で増大する。

$$
\gamma_{\mathrm{off}}
=
\gamma_{\mathrm{off}}(T,\Omega_{c1})
$$

が温度や optical power によって増大する。

あるいは、probe も隣接励起準位に結合し、

$$
\Omega_{p2}\neq0
$$

となるため、multi-$ \Lambda $ interference を明示的に入れる必要がある。

したがって、この notebook の Level 2 は、

$$
\mathrm{off-resonant\ light\ shift}
+
\mathrm{induced\ decoherence}
$$

を扱うモデルであり、さらに強い contrast 低下を再現したい場合には Level 3 の multi-$ \Lambda$ susceptibility へ進む必要があります。


## 9. パラメータを変えるときの注意

この notebook ではすべての周波数を MHz 単位で扱っています。

したがって、以下はすべて MHz で入力してください。

$$
\Omega_{c1},\ \Omega_{c2},\ \Delta_{\mathrm{off}},\ \gamma_{e_1g_2},\ \gamma_g,\ \gamma_{\mathrm{off}},\ \delta
$$

角周波数 \(\mathrm{rad/s}\) を使う場合は、すべての量に同じ規約を使ってください。  
一部だけ \(2\pi\) を含めると contrast の定量値が変わります。
