## Electromagnetically induced Transparency

Electromagnetically induced transparency (EIT) is a quantum interference effect that permits the propagation of light through an otherwise opaque atomic medium. The coherent optical nonlinearity renders a medium transparent within a narrow spectral range around an absorption line. Extreme dispersion is also created within this transparency "window" which leads to "slow light."

> Electromagnetically induced transparency (EIT) is an optical coherence effect which provides exquisite control over the absorption and dispersion in atomic media. In atoms with two coherent ground-state levels that can be optically excited to the same excited state (a “ 𝛬 system”), EIT results in ultranarrow transmissive spectral features, with resonance quality factors exceeding 10^12 and more than 10^7 reduction in optical group velocity. Numerous EIT-based applications are being pursued, including precision measurement, few-photon nonlinear optics, optical buffers, and quantum optical memories. --Wikipedia


EIT has been observed in atomic systems. The most notable platforms are rare-earth-doped systems. However, one drawback of these systems is weak optical transitions. As an alternative systemm, a solid-state approach is desirable for compatibility with large-scale fabrication processes.

# Research question and Hypothesis

## Sub issue 1: How does the lambda system of the ESLAC-induced NV-EIT differ from the strain-induced NV-EIT?

According to [Acosta 2013], the labmda system for the strain-induced NV-EIT consists of the $^3 A_2$ graound-state $m_s = 0$ and $m_s = +1$ and the excited states whose spin is mixed due to spin-orbit coupling. The past work only looked at the strain dependence of this coupling. This coupling also depends strongly on the applied magnetic field. Thus, we first calculate the ESLAC feature under various strain conditions by simply diagonalizing the Hamiltonian. We also try to find a sweet spot in the strain and magnetic field space, where two spin quantum numbers are mixed equally in an eigenstate. This is because we speculate that equal mixing of two spin numbers leads to higher EIT signal contrast.

### Hamiltonian definiton

Our initial goal is to understand the Hamiltonian of the system that we consider. It is well surmarized in the review work of [Doherty 2013].




#### The ground states

The ground-state bare Hamiltonian is

$$ \hat{H}_{gs} = D_{gs} \left[\hat{S}_z^2 - \frac{S(S+1)}{3}\right] + A_{gs}^{\parallel} \hat{S}_z \hat{I}_z + A_{gs}^{\perp}\left[\hat{S}_x\hat{I}_x+\hat{S}_y\hat{I}_y\right] + P_{gs}\left[\hat{I}_z^2 - \frac{I(I+1)}{3}\right],$$

where $D_{gs} = 2.87$ GHz is the zero-field splitting, $\hat{\vec{S}}= [\hat{S_x}, \hat{S_y}, \hat{S_z}]$ and $\hat{\vec{I}}= [\hat{I_x}, \hat{I_y}, \hat{I_z}]$ are the electronic and nuclear spin operators of $S = 1$ and $I=1$, respectively,\\ $A_{gs}^{\parallel}$ and $A_{gs}^{\perp}$ are the parallel and perpendicular components of the hyperfine tensor, and $P_{gs}$ is the quadrupolar coupling constant. Under the presence of a magnetic field $\vec{B} = [B_x, B_y, B_z]$, an electroc field $\vec{E} = [E_x, E_y, E_z]$, and a strain $\vec{\delta} = [\delta_x, \delta_y, \delta_z]$, the interaction Hamiltonian is  

$$ 
\begin{align}
\hat{V}_{gs} &= \mu_B g_{gs}^{\parallel} \hat{S}_z B_z + \mu_B g_{gs}^{\perp} \left(\hat{S}_x B_x + \hat{S}_y B_y \right) 
+\mu_Ng_N \hat{\vec{I}}\cdot \hat{\vec{S}} \\ \notag
&+ d_{gs}^{\parallel} (E_z+ \delta_z) \left[\hat{S_z}^2 - \frac{S(S+1)}{3} \right] + d_{gs}^{\perp} (E_x+ \delta_x) \left(\hat{S}_y^2 - \hat{S}_x^2 \right)+ d_{gs}^{\perp} (E_y+ \delta_y) \left(\hat{S}_x\hat{S}_y - \hat{S}_y\hat{S}_x \right),
\end{align}
$$

where $\mu_B$ is the Bohr magneton, $\mu_N$ is the nuclear magneton, $g_{gs}^{\parallel}$ and $g_{gs}^{\perp}$ are the components of the ground state electronic g-factor tensor, $g_N$ is the isotropic nuclear g-factor of 14N or 15N as required, and $d_{gs}^{\parallel}$ and $d_{gs}^{\perp}$ are the components of the ground state electric dipole moment.

### Numerical calculation of excited state energy levels

Our next task is to calculate the excited states as a function of magnetic field and strain by diagonalizing the total Hamiltonian. Their energy levels are obtained as the eigenvalues.

#### Program to diagonalize the Hamiltonian

To carry out this task, we use the qutip package. If qutip is not installed, or if you use the Google colab to run, please execute the following code block.

```python
# Execute this command if qutip is not installed in your python environment
!pip install qutip
```

Then we numerically defien the excited state Hamiltonian using the expression and coefficients introduced in the previous section.

```python
import qutip as qt
import numpy as np
import scipy as sp
import matplotlib
import matplotlib.pyplot as plt
#plt.rcParams['text.usetex'] = True
#rfont = {'fontname':'Times New Roman'}
matplotlib.rcParams.update(matplotlib.rcParamsDefault)

# Spin matrix
S0 = qt.qeye(3)                                                # {1, 0, -1}
Sx, Sy, Sz = qt.jmat(1)                                        # {1, 0, -1}
sigma0 = qt.qeye(2)                                            # {Ex, Ey}
sigmax, sigmay, sigmaz = qt.sigmax(), qt.sigmay(), qt.sigmaz() # {Ex, Ey}

# Physics constant
h             = 6.626e-34        # J/Hz
muB           = 9.274e-24 / h    # Hz/T
Des_para      = 1.42e9           # Hz       # [Doherty 2013], 1.44e9[Bassett 2014] 
Des_perp      = 1.55e9/2         # Hz       # [Doherty 2013], 1.541e9[Bassett 2014] 
lambdaes_para = 5.3e9            # Hz       # [Doherty 2013], 5.33e9[Bassett 2014]
lambdaes_perp = 0.2e9/np.sqrt(2) # Hz       # [Doherty 2013]
des_para      = 6e3              # Hz/(V/m) # [Doherty 2013]?
des_perp      = 6e3              # Hz/(V/m) # [Doherty 2013]?
les_para      = 0.1              # 1        # [Doherty 2013], 0.05[Bassett 2014] at 1.5K
ges_para      = 2.01             # 1        # [Doherty 2013], 2.15[Bassett 2014]
ges_perp      = 2.01             # 1        # [Doherty 2013]
```

The bare Hamiltonian and interaction Hamiltonian parts are numerically defined as follows.

```python
# bare Hamiltonian
H_es = Des_para * qt.tensor(sigma0, Sz**2 - 2/3 * S0) \
       - lambdaes_para * qt.tensor(sigmay, Sz) \
       + Des_perp * (qt.tensor(sigmaz, Sy**2 - Sx**2) - qt.tensor(sigmax, Sy * Sx + Sx * Sy)) \
       + lambdaes_perp * (qt.tensor(sigmaz, Sx * Sz + Sz * Sx) - qt.tensor(sigmax, Sy * Sz + Sz * Sy))

# Interaction Hamiltonian
def V_es(E, B, delta):
    return des_para * (E[2] + delta[2]) * qt.tensor(sigma0, S0) \
           + des_perp * (E[0] + delta[0]) * qt.tensor(sigmaz, S0) \
           - des_perp * (E[1] + delta[1]) * qt.tensor(sigmax, S0) \
           + muB * (les_para * qt.tensor(sigmay, S0) + ges_para * qt.tensor(sigma0, Sz)) * B[2] \
           + muB * ges_perp * qt.tensor(sigma0, Sx * B[0] + Sy * B[1])

# V_esの変数に電場、磁場、strain=delta が含まれている。そのため、励起状態におけるstrainの依存性がわかれば
# トータルのハミルトニアンを対角化することで、エネルギー準位を計算できそう (May 25 2025)
```

To obtain the energy levels under arbitrary electromagnetic fields and strain, one can use the qutip function "plot_energy_levels," which diagonalizes the Hamiltonian and creates a plot of the energy levels quickly.

#### Dependence of excited state energy on $\vec{B}$ and $\vec{\delta}$

Using the program defined in the previous section, we check how the excited state levels change as a function of magnetic field and strain.

##### Dependence on off-axis strain $\delta_{\perp} \equiv \sqrt{\delta_x^2 + \delta_y^2}$

We check how the excited state energy levels change according to the strain. By varing $d_{es}^{\perp} \delta_\perp$, we can reproduce the energy level diagram in [Doherty 2013][Maze 2011]. It's claer that the $|E_y⟩(0)$ and $|E_2⟩(+1)$ has a similar energy level near $d_{es}^{\perp} \delta_\perp = 8$ GHz point. The results calculated with the following code agree well with [Doherty 2013].

```python
# strain list
deltax = deltay = np.linspace(0, 20e9) / np.sqrt(2) / des_perp 

# eigenvalues
ee_strain = [(H_es + V_es(E=[0, 0, 0], B=[0, 0, 0], delta=[deltaxi, deltayi, 0])).eigenenergies() for deltaxi, deltayi in zip(deltax, deltay)]
#strainに応じて変化する全ハミルトニアン(変化するのはV_esだけど)に対応する固有値を、eigenenergies関数で計算
#今回は、ハミルトニアンのstrain依存性(NV軸からずれた場合？)のみをチェック

# plot
for i in range(6):
    plt.plot(des_perp * np.sqrt(deltax**2 + deltay**2)/1e9, [eei[i]/1e9 for eei in ee_strain]) #eei[i]を 1e9で割ってGHZに換算
plt.xlabel(r"Strain $d_{es}^{\perp} \delta_\perp \mathrm{(GHz)}$",fontsize=15)
plt.ylabel(r"Energy $E \mathrm{(GHz)}$",fontsize=15)
plt.tick_params(labelsize=15)
plt.show()
```

### Decomposition of excited eigenstates into spin quantumm basis

For the NV-EIT to work, a coherent mixing between $\Delta m_s = \pm 1$ needs to occur. For the strain-induced or ESLAC-induced EIT, we make sure that the excited state consists of multiple spin quantum states. Thus, we decompose the excited states into the spin quantum number basis.

The eigenstate basis is $\{|\psi_{es} \rangle \} = (|A_2\rangle, |A_1\rangle, |E_x\rangle, |E_y\rangle, |E_2\rangle, |E_1\rangle)$. The spin quantum number basis is $\{|\sigma_{es} \rangle \} = (|E_x, +1\rangle, |E_x, 0\rangle, |E_x, -1\rangle, |E_y, +1\rangle, |E_y, 0 \rangle, |E_y, -1\rangle)$.


#### Program to calculate the degree of mixture

The function "number_basis_tomography" in the next code block calculates the degree of mixture of the spin qunatum states for each excted eigenstate, i.e., $|\langle \sigma_{es} | \psi_{es} \rangle|^2$.

```python
def number_basis_tomography(H_tot):
    ee, ev = H_tot.eigenstates()
    ev_amp = [abs(evi.full())**2 for evi in ev][::-1]
    #配列の中に、各スピン状態の確率振幅の二乗を格納
    #インデックス::-1は不明


    xlabels = [r'$|E_x, 1\rangle$', r'$|E_x, 0\rangle$', r'$|E_x, -1\rangle$', r'$|E_y, 1\rangle$', r'$|E_y, 0\rangle$', r'$|E_y, -1\rangle$']
    ylabels = [r'$|E_2\rangle$', r'$|E_1\rangle$', r'$|E_y\rangle$', r'$|E_x\rangle$', r'$|A_1\rangle$', r'$|A_2\rangle$'][::-1]
    plt.imshow(ev_amp, cmap='Blues')
    plt.colorbar()
    plt.gca().xaxis.set_ticks_position('top')
    plt.xticks(ticks=np.arange(len(xlabels)), labels=xlabels)
    plt.yticks(ticks=np.arange(len(ylabels)), labels=ylabels)
    plt.show()

    return 0
```

### Numerical simulation of Coherent Population Trapping (work in progress)


Here, we are going to see if coherent population trapping happens at the conditions that we saw in the previous section.

#### Schrödinger equation construction
First, we need to construct the ground state triplet Hamiltonian without nuclear spin.

```python
# Physics constant
Dgs = 2.87e9                # Hz       # [Doherty 2013]
dgs_para = 0.035            # Hz/(V/m) # [Doherty 2013]
dgs_perp = 0.17             # Hz/(V/m) # [Doherty 2013]
ggs_para = 2.0028           # 1        # [Doherty 2013]
ggs_perp = 2.0028           # 1        # [Doherty 2013]
```

The ground state Hamiltonian is numerically defined as follows.

```python
# bare Hamiltonian
H_gs = Dgs * (Sz**2 - 1 * (1 + 1)/3 * S0)

# Interaction Hamiltonian
def V_gs(E, B, delta):
    return muB * ggs_para * Sz * B[2] \
           + muB * ggs_perp * (Sx * B[0] + Sy * B[1]) \
           + dgs_para * (E[2] + delta[2]) * (Sz**2 - 1 * (1 + 1)/3 * S0) \
           + dgs_perp * (E[0] + delta[0]) * (Sy**2 - Sx**2) \
           + dgs_perp * (E[1] + delta[1]) * (Sx * Sy + Sy * Sx)
```

We can check the ground state energy levels quickly.

```python
# plot energy levels
E = [0, 0, 0] # V/m
B = [0, 0, 100e-4] # T
delta = [0.75e10/des_para/np.sqrt(2), 0.75e10/des_para/np.sqrt(2), 0] # Hz

H_tot = H_gs + V_gs(E, B, delta)
qt.plot_energy_levels([H_tot], h_labels=None, energy_levels=None, N=0, fig=None, ax=None) #qutip 5.0
#qt.plot_energy_levels([H_tot], show_ylabels=True, figsize=(2, 3)) #qutip 4.0

plt.show()
```

Then, we combine ground state Hamiltonian and excited state Hamiltonian as follows.
$$
H = |^3A_2\rangle\langle^3A_2| \otimes H_{gs} + \omega_z |^3E\rangle\langle^3E| \otimes H_{es}
=
\begin{pmatrix}
H_{gs} & 0 \\
0 & \omega_z + H_{es}
\end{pmatrix}
$$

$$
V = |^3A_2\rangle\langle^3A_2| \otimes V_{gs} + |^3E\rangle\langle^3E| \otimes V_{es}
=
\begin{pmatrix}
V_{gs} & 0 \\
0 & V_{es}
\end{pmatrix}
$$

Here, we defined $\omega_z = c/ZPL = 470$ THz as a ZPL ramor freuquency

```python
# Define physics constant
c   = 3.0e8  # m/s
ZPL = 637e-9 # m

# full system bare Hamiltonian
H = np.zeros((9, 9), dtype=complex)
H[:3, :3] = H_gs.full()
H[3:, 3:] = H_es.full() + c/ZPL * np.eye(6)
H = qt.Qobj(H)

# full system interaction Hamiltonian
def V(E, B, delta):
    V = np.zeros((9, 9), dtype=complex)
    V[:3, :3] = V_gs(E, B, delta).full()
    V[3:, 3:] = V_es(E, B, delta).full()
    return qt.Qobj(V)
```

Then, we represent the incident laser as a complete spin-preserving transition between ground state and excited state.
$$
H_{laser} = P \cos (\omega t)\tilde{S}
$$

Here, we define the pseudo transition matrix as follows.
$$
\tilde{S} = (|^3A_2\rangle\langle^3E_x| + |^3A_2\rangle\langle^3E_y| + |^3E_x\rangle\langle^3A_2| + |^3E_y\rangle\langle^3A_2|) \otimes S_0
$$

```python
A2, Ex, Ey = qt.basis(3, 0), qt.basis(3, 1), qt.basis(3, 2)
S_pseudo = qt.Qobj(qt.tensor(A2 * Ex.dag() + A2 * Ey.dag() + Ex * A2.dag() + Ey * A2.dag(), S0).full())
```

Here we calculate the population dynamics by master equation in lab frame, as a function of $\vec{E}$, $\vec{B}$, $\vec{\delta}$, $\Omega_c$, $\lambda_c$, $\Omega_p$, $\lambda_p$, $t_{end}$

```python
def rho_CPT(E, B, delta, P1, lambda1, P2, lambda2, t_end): # E(V/m), B(T), delta(V/m), P(Hz), lambda(m), t_end(s)
  dt = 0.1e-12 # s
  t = np.linspace(0, t_end, int(t_end/dt) + 1)
  H0 = 2*np.pi * (H + V(E, B, delta))
  H1 = 2*np.pi * S_pseudo
  def H1_coeff(t, args):
    return P1 * np.cos(2*np.pi*c/lambda1 * t)
  H2 = 2*np.pi * S_pseudo
  def H2_coeff(t, args):
    return P2 * np.cos(2*np.pi*c/lambda2 * t)
  rho0 = 1/3 * (qt.ket2dm(qt.basis(9, 0)) + qt.ket2dm(qt.basis(9, 1)) + qt.ket2dm(qt.basis(9, 2)))
  result = qt.mesolve([H0, [H1, H1_coeff], [H2, H2_coeff]], rho0, t, options={'progress_bar': 'enhanced'})
  return t, result.states
```

## Rotating Wave Approximation

The calculation required a tremendous amount of time because we needed to use a time step of $\Delta t = 0.1$ ps. This is due to the fact that the ZPL at approximately $637$ nm corresponds to a Rabi precession frequency of about $470$ THz.  
To make the calculation more efficient, we employ a rotating frame, described by the following unitary matrix:

$$
\begin{aligned}
U(t) &= e^{i \omega_z |E\rangle\langle E| t} \\
&=
\begin{pmatrix}
I_3 & 0 \\
0 & e^{i\omega_z t} I_6
\end{pmatrix}.
\end{aligned}
$$

In the rotating frame, the Hamiltonian becomes:

$$
\begin{aligned}
H_{\mathrm{rot}} &= U H U^\dagger - i\, U \frac{dU^\dagger}{dt} \\
&=
\begin{pmatrix}
I_3 & 0 \\
0 & e^{i\omega_z t} I_6
\end{pmatrix}
\begin{pmatrix}
H_{\mathrm{gs}} & 0 \\
0 & \omega_z + H_{\mathrm{es}}
\end{pmatrix}
\begin{pmatrix}
I_3 & 0 \\
0 & e^{-i\omega_z t} I_6
\end{pmatrix}
-i
\begin{pmatrix}
I_3 & 0 \\
0 & e^{i\omega_z t} I_6
\end{pmatrix}
\frac{d}{dt}
\begin{pmatrix}
I_3 & 0 \\
0 & e^{-i\omega_z t} I_6
\end{pmatrix} \\
&\quad
+ P \cos(\omega_L t)
\begin{pmatrix}
I_3 & 0 & 0 \\
0 & e^{i\omega_z t} I_3 & 0 \\
0 & 0 & e^{i\omega_z t} I_3
\end{pmatrix}
\begin{pmatrix}
0 & I_3 & I_3 \\
I_3 & 0 & 0 \\
I_3 & 0 & 0
\end{pmatrix}
\begin{pmatrix}
I_3 & 0 & 0 \\
0 & e^{-i\omega_z t} I_3 & 0 \\
0 & 0 & e^{-i\omega_z t} I_3
\end{pmatrix} \\
&=
\begin{pmatrix}
H_{\mathrm{gs}} & 0 \\
0 & H_{\mathrm{es}}
\end{pmatrix}
+
\frac{P}{2}
\begin{pmatrix}
0 & e^{i(\omega_L - \omega_z)t}I_3 & e^{i(\omega_L - \omega_z)t}I_3 \\
e^{-i(\omega_L - \omega_z)t}I_3 & 0 & 0 \\
e^{-i(\omega_L - \omega_z)t}I_3 & 0 & 0
\end{pmatrix}.
\end{aligned}
$$


Note that we used rotating wave approximation here. The state vector in rotating frame becomes as follows.
$$
|\psi\rangle_{rot} = U|\psi\rangle
$$
The population dynamics wouldn't change in rotating frame because
$$
|\langle i|\psi\rangle_{rot}|^2 = \langle i|\psi_{rot}\rangle\langle \psi_{rot}|i\rangle = \langle i| U |\psi\rangle\langle\psi |U^\dagger|i\rangle =
$$

```python
def rho_CPT_RWA(E, B, delta, rho_init, P1, lambda1, P2, lambda2, t_end): # E(V/m), B(T), delta(V/m), P(Hz), lambda(m), t_end(s)
  # prepare
  c = 3e8 # m/s
  dt = 0.01e-9 # s
  t = np.linspace(0, t_end, int(t_end/dt) + 1)
  # Hamiltonian and its rotating wave approximation
  H0 = 2*np.pi * (H + V(E, B, delta))
  H0_rot = qt.Qobj(H0.full() - 2*np.pi*c/ZPL*np.diag([0, 0, 0, 1, 1, 1, 1, 1, 1]))
  delta_omega1 = 2*np.pi*c*(1/lambda1 - 1/ZPL)
  delta_omega2 = 2*np.pi*c*(1/lambda2 - 1/ZPL)
  H_up_rot = qt.Qobj(np.kron(np.array([[0, 1, 1], [0, 0, 0], [0, 0, 0]]), np.eye(3)))
  def H_up_rot_coeff(t, args):
    return P1/2 * np.exp(1j*delta_omega1*t) + P2/2 * np.exp(1j*delta_omega2*t)
  H_down_rot = qt.Qobj(np.kron(np.array([[0, 0, 0], [1, 0, 0], [1, 0, 0]]), np.eye(3)))
  def H_down_rot_coeff(t, args):
    return P1/2 * np.exp(-1j*delta_omega1*t) + P2/2 * np.exp(-1j*delta_omega2*t)
  result = qt.mesolve([H0_rot, [H_up_rot, H_up_rot_coeff], [H_down_rot, H_down_rot_coeff]], rho_init, t, options={'progress_bar': 'enhanced', 'rtol': 1e-8, 'atol': 1e-8})
  return t, result.states
```

## Sub issue 3: At which temperature does the ESLAC induced NV-EIT break down?

In this issue 3, we would like to see how robust the EIT signal is against temperature. To carry out this task, we need to use the Master equation to include the phonon effect. Because a recent paper by [Happacher 2023] provides a full description for predicting photoluminescence under intermediate temperature, we first try to reproduce their results of photoluminescence under various temperature and magnetic field conditions. Then, we can incorporate the resonant driving of the lambda system to predict the EIT signal under intermediate temperature.

#### Master equation incorporating phonon interactions

Here, we are going to numerically calculate the temperature dependence of ESLAC induced NV-EIT. The simulation tool consists of three parts. The first part is the bare Hamiltonian of $^3A_2$, $^3E$, and $^1A_1 / E_1$, as well as their interaction Hamiltonian. The number of eigenstates is $3 + 6 + 1 = 10$. The Hamiltonians are from [Doherty 2013]. The second part is the spin preserving transitions between $^3A_2$, $^3E$, and $^1A_1 / E_1$. These transition include laser pumping, spontaneous optical decay, and inter sytem crossing. The transitions are expressed as Lindblad operator, whose decay rate is independent on temperature except for the inter sytem crossing rate. The decay rates are from [Happacher 2023]. The third part is the phonon-induced transitions within excited state. The trantision matrices are basically expressed by only $|E_x\rangle$ and $|E_y\rangle$, according to group thoery [Happacher 2023]. Their rates have a strong temperature dependence of $\propto T^5$ [Happacher 2023].

We expect to have the temperature hypothesis as following steps.
1. Bulid the model described above
1. Prepare initial state on the excited state that spin mixing happens
1. Solve the master equation and have coherence time of that state
1. Compare the coherence time with low temparature case

These steps are based on the following assumings.
- EIT robustness is a function of
  - Incident laser linewidth and power
  - Excited state eigenstate's spin states balance that we are using
  - Spontaneous decay rates
  - ISC decay rates
  - Photon induced transition within excited states
  - T1 at ground states
- Laser linewidth is very narrow enough to make transition only of our interest states
- We can change laser power as we like
- Excited state eigenstate's spin states balance are optimized to 50-50
- Spontaneous decay rates are inpependent on temperature
- T1 at ground states are independent on temperature

Second, define laser pumping and spontaneous decay and ISC. Let us define temperature independent rate first.

```python
# Temperature independent rate
k_r    = 66.8e6   # Hz # spontaneous decay  # [Happacher 2023]
beta_x = 0.0634   # 1  # pumping parameter  # [Happacher 2023]
beta_y = 0.039    # 1  # pumping parameter  # [Happacher 2023]
k_nr1  = 90.7e6   # Hz # Excited to Singlet # [Happacher 2023]
k_nr0  = 10.5e6   # Hz # Excited to Singlet # [Happacher 2023]
```

The ISC rate from Singlet to Ground state follows Bose-Einstein statistics as follows.
$$
\tau (T) = \tau_0 [1 - \exp \left(\frac{-\Delta E}{k_B T}\right)]
$$
$$
\frac{1}{\tau (T)} = k_{m_0} + k_{m_{\pm 1}} = k_{m_0} + k_{m_{-1}} + k_{m_1}
$$
The photon energy $\Delta E = 16.6$ meV and $\tau_0 = 296.53$ ns. The branching ratio is $k_{m_0} : k_{m_{-1}} : k_{m_1} = 4.36 : 1 : 1$ [Happacher 2023]

```python
# Temperature dependent rate, ISC
def tau(T):
  tau_0   = 296.53e-9 # s    # [Happacher 2023]
  Delta_E = 16.6e-3   # eV   # [Happacher 2023]
  k_B     = 8.617e-5  # eV/K # [Happacher 2023]
  return tau_0 * (1 - np.exp(-Delta_E / (k_B * T)))

def k_m0(T):
  return 4.36/(4.36 + 1 + 1) * 1/tau(T)

def k_m1(T):
  return 1/(4.36 + 1 + 1) * 1/tau(T)
```

The photon-induced transition probability can be derived from electron-phonon interaction Hamiltonian. Let us skip the deriviation for now(Need to check any assumption or approximation in it). The result is as follows.
$$
k_{|X\rangle →|Y\rangle} = k_{|Y\rangle → |X\rangle} = \alpha_{ph} I(\omega_c = 60 \mathrm{meV}, T, \delta_\perp) T^5
$$
The phonon coupling parameter is $\alpha_{ph} = 1.4$ $\mathrm{K^5 Hz}$. The $I(\omega_c, T, \delta_\perp)$ is an integral part which corresponds to the summation part of Fermi's golden rule as follows.
$$
I(\omega_c, T, \delta_\perp) = \int^{x_c}_0 \mathrm{dx} x (x + x_\Delta)[x^2 + (x + x_\Delta)^2]\frac{1}{(e^x - 1)(1 - e^{-(x + x_\Delta)})}
$$
with $x_c = \hbar \omega_c / (k_B T)$, $\omega_c$ being the cutoff frequency in the ingegral, $x = \hbar \omega / (k_B T)$, and $x_\Delta = 2\sqrt{\delta_x^2 + \delta_y^2} / (k_B T)$.

```python
# Temperature dependent rate, Photon-induced transition probability
alpha_ph = 1.4 # Hz/K^5

delta = [10e9/des_perp/np.sqrt(2), 10e9/des_perp/np.sqrt(2), 0] # V/m

def I(omega_c, T, delta): # omega_c (eV), T (K), delta (V/m)
  h       = 4.135e-15                                                         # eV s
  print(omega_c/h)
  k_B     = 8.617e-5                                                          # eV/K
  x_c     = omega_c / (k_B * T)                                               # 1
  x_delta = h * 2 * des_perp * np.sqrt(delta[0]**2 + delta[1]**2) / (k_B * T) # 1
  x       = np.linspace(x_c/100*1, 0.9*x_c, 100)                              # 1
  dx      = x[1] - x[0]                                                       # 1
  return np.sum(dx * x * (x + x_delta) * (x**2 + (x + x_delta)**2) / (np.exp(x) - 1) / (1 - np.exp(-(x + x_delta))))

def k_xy(T, delta):
  return alpha_ph * I(60e-3, T, delta) * T**5 # hbar * omega_c = 60e-6 eV [Happacher 2023]

val = float(k_xy(100,delta))
print(val/1e9)
```

##### Superoperator
Here, we'd like to use superoperator formalism for efficient calculation.

The vectorlized density matrix is as follows.
$$
\hat{\rho} = \sum_{i, j} \rho_{i, j} |j\rangle|i\rangle
$$

```python
def dm2super(rho): # input:Qobj, output:numpy_array
  n = rho.shape[0]
  rho_super = qt.tensor(qt.zero_ket(n), qt.zero_ket(n))
  for i in range(n):
    for j in range(n):
      rho_super += rho[i, j] * qt.tensor(qt.basis(n, j), qt.basis(n, i))
  return rho_super.full()
```

Lindbladian superoperator is giben by as follows.
$$
\hat{\mathcal{L}} = i(\bar{H} \otimes I_{10} - I_{10} \otimes \bar{H}) + \sum_l \left( \bar{L_l} \otimes L_l - \frac{1}{2} I_{10} \otimes L_l^\dagger L_l - \frac{1}{2} \bar{L_l}^\dagger \bar{L_l} \otimes I_{10}\right)
$$


Linbladian 方程式
https://shyphys.org/memorandum/lindblad.html

super operatorについて
https://whyitsso.net/physics/quantum_mechanics/Lindblad_ex.html

```python
def HL2super(H, Llist): # input:Qobj, output:numpy_array
  H = H.full()
  n = H.shape[0]
  H_super = 1j * (np.kron(np.conjugate(H), np.eye(n)) - np.kron(np.eye(n), H))
  L_super = np.zeros([n**2, n**2], dtype=np.complex128)
  for L in Llist:
    L = L.full()
    L_super += np.kron(np.conjugate(L), L) - 1/2 * np.kron(np.eye(n), np.conjugate(L).T @ L) - 1/2 * np.kron(np.conjugate(L).T @ np.conjugate(L), np.eye(n))
  return H_super + L_super
```

The time-evolution of the system in the vectorized form is given by as follows.
$$
\frac{d\hat{\rho}(t)}{dt} = \hat{\mathcal{L}}\hat{\rho}(t)
$$
$$
\hat{\rho}(t) = e^{\mathcal{L}t}\hat{\rho}(0)
$$
After the calculation, the superoperator density matrix is converted back to its matrix form.

```python
def super2dm(rho_super): # input:numpy_array, output:Qobj
  n = int(np.sqrt(rho_super.shape[0]))
  rho = np.zeros((n, n), dtype=complex)
  for i in range(n):
    for j in range(n):
      rho[i, j] = rho_super[i*n + j, 0]
  return qt.Qobj(rho)
```

Let us see the population dynamics which we saw in master eqation matrix form previously.

```python
# Variables
E = [0, 0, 0] # V/m
B = [0, 0, 0] # T
delta = [10e9/des_perp/np.sqrt(2), 10e9/des_perp/np.sqrt(2), 0] # V/m
T = 3 # K
t_end = 3e-6 # s
rho_init = rho_0

# Convert matrix to superoperator
rho_init_super = dm2super(rho_init)
HL_super = HL2super(H + V(E, B, delta), L(T, delta,with_singlet=True))

# Calculation
tlist_super = np.linspace(0, t_end, int(t_end/1e-9) + 1) # 1 ns step
rhos_super = [sp.linalg.expm(HL_super * t) @ rho_init_super for t in tlist_super]

# Convert superoperator to matrix
rhos_super = [super2dm(rho_super) for rho_super in rhos_super]

# Post processing
pop_super = [[qt.expect(pop_op, r) for r in rhos_super] for pop_op in pop_ops]
for i in range(len(pop_super)):
    plt.plot(tlist_super, pop_super[i], label=str(i+1))
plt.xlabel('Time (s)',fontsize=18)
plt.ylabel('Population',fontsize=18)
plt.legend()
plt.tick_params(labelsize=15)

plt.show()
```

#### Definition of photoluminescence

The photoluminescence is defined by the population in $^3 E$, multiplied by the decay rate $k_r$:
$$
I_{PL} = \sum_{i = 4\sim9} k_r p_i + I_{bck}.
$$
Here, $p_i$ is the population in state of $|i\rangle$, while $I_{bck} = 18$ kcps is the background fluorescensce.

```python
def I_PL(pop_ss):
  eta_collection = 0.03 # collection efficiency
  I_bck = 18e3 # cps

  return eta_collection * k_r * (pop_ss[3] + pop_ss[4] + pop_ss[5] + pop_ss[6] + pop_ss[7] + pop_ss[8]) + I_bck
```

### Does the NV-EIT signal appear as a peak or a dip? -> Dip

In this experiment, we detect a transmitted and a fluorescent light on top of the diamond sample. Let $I_0$ be the incident red laser power, $I_{A}$ the absorbed red laser power, $I_{T}$ the transimtted red laser power, and $I_{F}$ the red fluorescence collected by the detector. The red power into the detector $I_D$ is given by

$$I_D = I_T + I_F. $$

The transmitted power is $I_T = I_0 - I_A$, and the absorbed power is $I_A(\alpha) = I_0\left[1- \exp{(-\alpha z)}\right]$, where $\alpha$ is the absorption coefficient and $z$ is the path length. The fluorescnce detected is proportional to the absorbed light and the colelction efficiency: $I_F = \beta I_A$, where $\beta$ is the proportionality constant that includes the effect of collection efficiency. We assume that $I_0$, $z$, and $\beta$ stay constant, while $\alpha$ changes its value depending on the resonance. Then the detected red power is

$$I_D(\alpha) = I_0 - I_A(\alpha) + \beta I_A(\alpha) = I_0  -I_0\left[1-\exp{(-\alpha z)}\right](1-\beta)$$

Let $\alpha_{on}$ and $\alpha_{off}$ be the absorption coefficients when the frequency of the red laser is tuned on resonance and off resonance. Then the red signal contrast $C$ is given by

$$ C = \frac{I_D(\alpha_{on})}{ I_D(\alpha_{off})} = \frac{1-\left[1-\exp{(-\alpha_{on} z)}\right](1-\beta)}{1-\left[1-\exp{(-\alpha_{off} z)}\right](1-\beta)}.$$


Because the absorption rate increases for the resonance, we can say $\alpha_{on} \geq \alpha_{off} $. Then the following quick numerical calculation suggests that the signal is a dip.

```python
import numpy as np

a_on = 0.12
a_off = 0.1
z = 10
beta  = 0.05
I0 = 1

Ion = I0 - I0 *(1-np.exp(-a_on *z))*(1-beta)
Ioff = I0 - I0 *(1-np.exp(-a_off *z))*(1-beta)
C = Ion / Ioff
print(C)
```
