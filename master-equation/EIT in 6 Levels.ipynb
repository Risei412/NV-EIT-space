                    - \frac{\gamma_{2 \, \mathrm{deph}}}{2} \rho_{21}$

#### EITの導出

EITは次の式で与えられる

$\mathrm{Im} \left[\chi \left(-\omega_{\mathrm{p}}, +\omega_{\mathrm{p}} \right) \right] = \mathrm{Im} \left( \rho_{13} \mu_{31} 
                                    + \rho_{23} \mu_{32} + \rho_{24} \mu_{42} + \mathrm{C.c}  \right) $

このうち、Probe光に関するものだけを見ると、

$\mathrm{Im} \left[\chi \left(-\omega_{\mathrm{p}}, +\omega_{\mathrm{p}} \right) \right] = \mathrm{Im} \left( \rho_{13} \mu_{31} 
                                    + \rho_{23} \mu_{32} + \rho_{24} \mu_{42} + \mathrm{C.c}  \right) $

解いてみると、下記の式を得る(Gemini pro より)

$$\mathrm{Im} [\chi] \propto \mathrm{Im} \left\{ -\frac{\mathrm{i}}{2} \left( \frac{\mu_{31}}{\gamma_{31}} \left[ \Omega_{\mathrm{p},1}^* \left( \rho_{11} - \rho_{33} \right) - \Omega_{\mathrm{p,2}}^* \rho_{43} + \Omega_{\mathrm{c},1}^* \rho_{12} \right] \\ + \frac{\mu_{32}}{\gamma_{32}} \left[ \Omega_{\mathrm{c},1}^* \left( \rho_{22} - \rho_{33} \right) - \Omega_{\mathrm{c,2}}^* \rho_{43} + \Omega_{\mathrm{p},1}^* \rho_{12} \right] \\ + \frac{\mu_{42}}{\gamma_{42}} \left[ \Omega_{\mathrm{c},2}^* \left( \rho_{22} - \rho_{44} \right) - \Omega_{\mathrm{c,1}}^* \rho_{43} + \Omega_{\mathrm{p},2}^* \rho_{12} \right] \right) \right\}$$

```python
# constants
des_perp = 6e3      # Hz/(V/m)
alpha_ph = 1.4      # Hz/K^5

def I(omega_c, T, delta):  # omega_c [eV], T [K], delta [V/m]
    h   = 4.135e-15        # eV s
    k_B = 8.617e-5         # eV/K

    x_c = omega_c / (k_B * T)

    delta_abs = np.sqrt(delta[0]**2 + delta[1]**2)
    x_delta = h * 2 * des_perp * delta_abs / (k_B * T)

    x = np.linspace(x_c / 100, 0.9 * x_c, 100)
    dx = x[1] - x[0]

    integrand = (
        x
        * (x + x_delta)
        * (x**2 + (x + x_delta)**2)
        / (np.exp(x) - 1)
        / (1 - np.exp(-(x + x_delta)))
    )

    return np.sum(dx * integrand)


def plot_I_heatmap(
    omega_c=60e-3,
    T_min=10,
    T_max=300,
    strain_min_GHz=0,
    strain_max_GHz=70,
    n_T=200,
    n_strain=200,
    log_scale=True,
):
    """
    I(omega_c, T, delta) のヒートマップを描画する関数。

    横軸: Temperature T [K]
    縦軸: perpendicular crystal strain splitting Δ_perp [GHz]

    Δ_perp = 2 * des_perp * sqrt(delta_x^2 + delta_y^2)
    として、GHz -> delta [V/m] に変換する。
    """

    T_list = np.linspace(T_min, T_max, n_T)
    strain_GHz_list = np.linspace(strain_min_GHz, strain_max_GHz, n_strain)

    I_map = np.zeros((n_strain, n_T))

    for i, strain_GHz in enumerate(strain_GHz_list):
        strain_Hz = strain_GHz * 1e9

        # Δ_perp = 2 * des_perp * |delta_perp|
        delta_abs = strain_Hz / (2 * des_perp)

        # x方向だけにひずみを入れる表現
        delta_vec = [delta_abs, 0.0, 0.0]

        for j, T in enumerate(T_list):
            I_map[i, j] = I(omega_c, T, delta_vec)

    if log_scale:
        plot_data = np.log10(I_map + 1e-300)
        cbar_label = r"$\log_{10} I$"
    else:
        plot_data = I_map
        cbar_label = r"$I$"

    plt.figure(figsize=(7, 5))

    extent = [
        T_min,
        T_max,
        strain_min_GHz,
        strain_max_GHz,
    ]

    im = plt.imshow(
        plot_data,
        origin="lower",
        aspect="auto",
        extent=extent,
    )

    plt.colorbar(im, label=cbar_label)
    plt.xlabel("Temperature T [K]")
    plt.ylabel(r"Perpendicular crystal strain $\Delta_\perp$ [GHz]")
    plt.title(r"Heatmap of $I(\omega_c, T, \Delta_\perp)$")

    plt.tight_layout()
    plt.show()

    return T_list, strain_GHz_list, I_map
```

```python
import numpy as np
import matplotlib.pyplot as plt
from dataclasses import dataclass


# ============================================================
# NV-EIT model:
#
# |1> = ground state |g>
# |2> = metastable / spin ground state |s>
# |3> = main excited state |e>
# |4> = extra excited state |e'>
#
# Probe:
#   |1> <-> |3> : Omega_p1
#   |1> <-> |4> : Omega_p2
#
# Control:
#   |2> <-> |3> : Omega_c1
#   |2> <-> |4> : Omega_c2
#
# Phonon-induced transition:
#   |3> -> |4> : k34
#   |4> -> |3> : k43
#
# All frequencies and rates are in MHz.
# ============================================================


@dataclass
class NVEITParams:
    Omega_p1: float = 0.01      # weak probe Rabi frequency
    Omega_p2: float = 0.0       # probe coupling to extra excited state
    Omega_c1: float = 20.0      # control coupling to main excited state
    Omega_c2: float = 0.0       # control leakage to extra excited state

    Delta_c: float = 0.0        # control detuning
    Delta_e: float = 80.0       # energy separation between |3> and |4>

    gamma31_0: float = 3.0      # optical coherence decay for |3>-|1>
    gamma41_0: float = 3.0      # optical coherence decay for |4>-|1>
    gamma21: float = 0.01       # ground-state coherence decay

    k34: float = 0.0            # phonon transition |3> -> |4>
    k43: float = 0.0            # phonon transition |4> -> |3>

    mu31: float = 1.0           # relative dipole moment |1>-|3>
    mu41: float = 1.0           # relative dipole moment |1>-|4>


def solve_chi_nv(delta_p, p: NVEITParams):
    """
    Calculate linear susceptibility chi for one probe detuning delta_p.

    Solves the steady-state linear equations:
        G31*rho31 - i*Omega_c1/2*rho21 = i*Omega_p1/2
        G41*rho41 - i*Omega_c2/2*rho21 = i*Omega_p2/2
        G21*rho21 - i*Omega_c1*/2*rho31 - i*Omega_c2*/2*rho41 = 0

    chi is normalized by Omega_p1:
        chi ~ (mu31*rho31 + mu41*rho41) / Omega_p1
    """

    # Phonon-induced transitions increase the outgoing decoherence
    gamma31_eff = p.gamma31_0 + 0.5 * p.k34
    gamma41_eff = p.gamma41_0 + 0.5 * p.k43

    delta_two = delta_p - p.Delta_c

    G31 = gamma31_eff - 1j * delta_p
    G41 = gamma41_eff - 1j * (delta_p - p.Delta_e)
    G21 = p.gamma21 - 1j * delta_two

    A = np.array(
        [
            [G31, 0.0 + 0.0j, -1j * p.Omega_c1 / 2],
            [0.0 + 0.0j, G41, -1j * p.Omega_c2 / 2],
            [-1j * np.conj(p.Omega_c1) / 2,
             -1j * np.conj(p.Omega_c2) / 2,
             G21],
        ],
        dtype=complex,
    )

    b = np.array(
        [
            1j * p.Omega_p1 / 2,
            1j * p.Omega_p2 / 2,
            0.0 + 0.0j,
        ],
        dtype=complex,
    )

    rho31, rho41, rho21 = np.linalg.solve(A, b)

    chi = (p.mu31 * rho31 + p.mu41 * rho41) / p.Omega_p1

    return chi, rho31, rho41, rho21


def spectrum(delta_p_array, p: NVEITParams):
    """
    Calculate chi and absorption spectrum.
    Absorption is defined as Im[chi].
    """

    chi_list = []
    rho31_list = []
    rho41_list = []
    rho21_list = []

    for dp in delta_p_array:
        chi, rho31, rho41, rho21 = solve_chi_nv(dp, p)
        chi_list.append(chi)
        rho31_list.append(rho31)
        rho41_list.append(rho41)
        rho21_list.append(rho21)

    chi_array = np.array(chi_list)
    absorption = np.imag(chi_array)

    return {
        "delta_p": np.array(delta_p_array),
        "chi": chi_array,
        "absorption": absorption,
        "rho31": np.array(rho31_list),
        "rho41": np.array(rho41_list),
        "rho21": np.array(rho21_list),
    }


def eit_contrast(delta_p_array, absorption, center_window=5.0):
    """
    Estimate EIT contrast.

    baseline:
        average absorption at the outer 10% of detuning range

    EIT minimum:
        minimum absorption around delta_p = 0

    contrast:
        (baseline - absorption_min) / baseline
    """

    delta_p_array = np.array(delta_p_array)
    absorption = np.array(absorption)

    max_detuning = np.max(np.abs(delta_p_array))
    edge_mask = np.abs(delta_p_array) > 0.8 * max_detuning
    center_mask = np.abs(delta_p_array) < center_window

    baseline = np.mean(absorption[edge_mask])
    absorption_min = np.min(absorption[center_mask])

    if np.abs(baseline) < 1e-15:
        return np.nan

    contrast = (baseline - absorption_min) / baseline
    return contrast


def plot_spectra():
    """
    Compare:
    1. Ideal 3-level EIT
    2. Extra excited state with Omega_c2
    3. Extra excited state + phonon transition
    """

    delta_p = np.linspace(-120, 160, 2001)

    # Case 1: ideal 3-level EIT
    p_3level = NVEITParams(
        Omega_p1=0.01,
        Omega_p2=0.0,
        Omega_c1=20.0,
        Omega_c2=0.0,
        Delta_e=80.0,
        gamma31_0=3.0,
        gamma41_0=3.0,
        gamma21=0.01,
        k34=0.0,
        k43=0.0,
    )

    # Case 2: extra excited state, coherent leakage only
    p_extra = NVEITParams(
        Omega_p1=0.01,
        Omega_p2=0.0,
        Omega_c1=40.0,
        Omega_c2=20.0,
        Delta_e=0.0,
        gamma31_0=10,
        gamma41_0=10,
        gamma21=0.1,
        k34=0.0,
        k43=0.0,
    )

    # Case 3: extra excited state + phonon transition
    p_phonon = NVEITParams(
        Omega_p1=0.01,
        Omega_p2=0.0,
        Omega_c1=50.0,
        Omega_c2=20.0,
        Delta_e=80.0,
        gamma31_0=3.0,
        gamma41_0=3.0,
        gamma21=0.01,
        k34=30.0,
        k43=30.0,
    )

    s1 = spectrum(delta_p, p_3level)
    s2 = spectrum(delta_p, p_extra)
    s3 = spectrum(delta_p, p_phonon)

    c1 = eit_contrast(delta_p, s1["absorption"])
    c2 = eit_contrast(delta_p, s2["absorption"])
    c3 = eit_contrast(delta_p, s3["absorption"])

    plt.figure(figsize=(7, 4.5))
    plt.plot(delta_p, s1["absorption"], label=f"3-level, contrast={c1:.3f}")
    plt.plot(delta_p, s2["absorption"], label=f"extra level, contrast={c2:.3f}")
    plt.plot(delta_p, s3["absorption"], label=f"extra level + phonon, contrast={c3:.3f}")
    plt.axvline(0, linestyle="--", linewidth=1)
    plt.axvline(p_3level.Delta_e, linestyle="--", linewidth=1)
    plt.xlabel(r"Probe detuning $\Delta_p$ (MHz)")
    plt.ylabel(r"Absorption $\mathrm{Im}[\chi_{\mathrm{NV}}]$")
    plt.title("NV-EIT with extra excited state and phonon transition")
    plt.legend()
    plt.tight_layout()
    plt.show()


def contrast_heatmap():
    """
    Plot EIT contrast as a function of:
        Omega_c2 / Delta_e
        k_phonon

    This shows the condition where EIT contrast collapses.
    """

    delta_p = np.linspace(-80, 80, 1201)

    Delta_e = 80.0
    Omega_c2_values = np.linspace(0.0, 120.0, 80)
    k_values = np.linspace(0.0, 120.0, 80)

    contrast_map = np.zeros((len(k_values), len(Omega_c2_values)))

    for i, k in enumerate(k_values):
        for j, Omega_c2 in enumerate(Omega_c2_values):
            p = NVEITParams(
                Omega_p1=0.01,
                Omega_p2=0.0,
                Omega_c1=20.0,
                Omega_c2=Omega_c2,
                Delta_e=Delta_e,
                gamma31_0=3.0,
                gamma41_0=3.0,
                gamma21=0.01,
                k34=k,
                k43=k,
            )

            s = spectrum(delta_p, p)
            contrast_map[i, j] = eit_contrast(
                delta_p,
                s["absorption"],
                center_window=5.0,
            )

    X, Y = np.meshgrid(Omega_c2_values / Delta_e, k_values)

    plt.figure(figsize=(6.5, 5))
    plt.pcolormesh(X, Y, contrast_map, shading="auto")
    plt.xlabel(r"$\Omega_{c2}/\Delta_e$")
    plt.ylabel(r"$k_{\mathrm{phonon}}$ (MHz)")
    plt.title("EIT contrast map")
    plt.colorbar(label="EIT contrast")
    plt.tight_layout()
    plt.show()


if __name__ == "__main__":
    plot_spectra()
    # contrast_heatmap()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def solve_eit_4level_approx(
    Delta_p_array,
    Omega_p1,
    Omega_p2,
    Omega_c1,
    Omega_c2,
    gamma31,
    gamma41,
    gamma21,
    Delta_e=0.0,
    Delta_c=0.0,
    mu31=1.0,
    mu41=1.0,
):
    """
    4準位近似EIT（弱プローブ・定常状態）の線形解を返す

    Parameters
    ----------
    Delta_p_array : 1D array
        プローブ detuning 掃引
    Omega_p1, Omega_p2 : complex
        |1>-|3>, |1>-|4> に対するプローブRabi
    Omega_c1, Omega_c2 : complex
        |2>-|3>, |2>-|4> に対するコントロールRabi
    gamma31, gamma41, gamma21 : float or complex
        有効緩和（単位はすべて統一）
        例:
          gamma31 = (Gamma31 + Gamma32 + gamma3_deph + k35) / 2
          gamma41 = (Gamma41 + Gamma42 + gamma4_deph + k46) / 2
          gamma21 = gamma2_deph / 2
    Delta_e : float
        excited state |4> が |3> からどれだけ離れているか
        例: Delta41 = Delta_p - Delta_e
    Delta_c : float
        control の detuning（|2>-|3>基準）
    mu31, mu41 : float or complex
        双極子モーメント（相対値で十分）

    Returns
    -------
    result : dict
        rho31, rho41, rho21, chi, absorption, dispersion
    """

    rho31_list = []
    rho41_list = []
    rho21_list = []
    chi_list = []

    for Delta_p in Delta_p_array:
        # detuning の定義
        Delta31 = Delta_p
        Delta41 = Delta_p - Delta_e
        delta_two_photon = Delta_p - Delta_c

        G31 = gamma31 - 1j * Delta31
        G41 = gamma41 - 1j * Delta41
        G21 = gamma21 - 1j * delta_two_photon

        # A x = b
        # x = [rho31, rho41, rho21]
        A = np.array([
            [G31, 0.0, -1j * Omega_c1 / 2.0],
            [0.0, G41, -1j * Omega_c2 / 2.0],
            [1j * np.conj(Omega_c1) / 2.0,
             1j * np.conj(Omega_c2) / 2.0,
             G21]
        ], dtype=complex)

        b = np.array([
            1j * Omega_p1 / 2.0,
            1j * Omega_p2 / 2.0,
            0.0
        ], dtype=complex)

        rho31, rho41, rho21 = np.linalg.solve(A, b)

        # プローブ周波数成分の susceptibility
        # 位相規約によっては conj を取る方が合う場合があります
        chi_p = mu31 * rho31 + mu41 * rho41

        rho31_list.append(rho31)
        rho41_list.append(rho41)
        rho21_list.append(rho21)
        chi_list.append(chi_p)

    rho31_arr = np.array(rho31_list)
    rho41_arr = np.array(rho41_list)
    rho21_arr = np.array(rho21_list)
    chi_arr = np.array(chi_list)

    result = {
        "rho31": rho31_arr,
        "rho41": rho41_arr,
        "rho21": rho21_arr,
        "chi": chi_arr,
        "absorption": np.imag(chi_arr),
        "dispersion": np.real(chi_arr),
    }
    return result


if __name__ == "__main__":
    # 単位は全部そろえること
    # ここでは便宜上 MHz をそのまま使う
    MHz = 1.0

    # probe detuning sweep
    Delta_p = np.linspace(-400, 400, 4001) * MHz

    # 例のパラメータ

    Omega_c1 = 600* MHz
    Omega_c2 = 0.8 * Omega_c1
    Omega_p1 = 0.08 * Omega_c1
    Omega_p2 = 0.8 * Omega_p1   # |4> への leakage を見たいなら 0 にしない

    gamma31 = 1000 * MHz
    gamma41 = 1000 * MHz
    gamma21 = 1 * MHz

    Delta_e = 10.0 * MHz     # |4> と |3> の差
    Delta_c = 0.0 * MHz     # control を |2>-|3> に共鳴させるなら 0

    out = solve_eit_4level_approx(
        Delta_p_array=Delta_p,
        Omega_p1=Omega_p1,
        Omega_p2=Omega_p2,
        Omega_c1=Omega_c1,
        Omega_c2=Omega_c2,
        gamma31=gamma31,
        gamma41=gamma41,
        gamma21=gamma21,
        Delta_e=Delta_e,
        Delta_c=Delta_c,
        mu31=1.0,
        mu41=1.0,
    )

    abs_sig = out["absorption"]

    # 見やすく正規化
    abs_norm = abs_sig / np.max(np.abs(abs_sig))

    plt.figure(figsize=(7, 4))
    plt.plot(Delta_p / MHz, abs_norm, label="Approx. EIT signal")
    plt.axvline(0, ls="--", alpha=0.5)
    plt.xlabel("Probe detuning (arb. unit or MHz)")
    plt.ylabel("Im[chi] (normalized)")
    plt.title("Approximate EIT spectrum with extra excited state |4>")
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


def solve_eit_4level_leakage(
    Delta_p_array,
    Omega_p1,
    Omega_c1,
    r_p=0.0,
    r_c=0.0,
    gamma31=1000.0,
    gamma41=1000.0,
    gamma21=1.0,
    Delta_e=500.0,
    Delta_c=0.0,
    mu31=1.0,
    mu41=None,
):
    """
    近接励起準位 |4> を含む4準位EIT線形応答モデル。

    States:
        |1> --probe--> |3>
        |2> --control-> |3>

        |1> --probe leakage--> |4>
        |2> --control leakage-> |4>

    Important:
        Omega_p2 = r_p * Omega_p1
        Omega_c2 = r_c * Omega_c1

    r_p and r_c are determined by transition dipoles, polarization,
    and Clebsch-Gordan-like coefficients.

    If r_p == r_c, a common dark state can still exist.
    If r_p != r_c, the dark-state cancellation is incomplete.
    """

    if mu41 is None:
        # probe susceptibility contribution should use the same
        # relative dipole as the probe Rabi coupling.
        mu41 = r_p

    Omega_p2 = r_p * Omega_p1
    Omega_c2 = r_c * Omega_c1

    rho31_list = []
    rho41_list = []
    rho21_list = []
    chi_list = []

    for Delta_p in Delta_p_array:
        Delta31 = Delta_p
        Delta41 = Delta_p - Delta_e
        delta_two_photon = Delta_p - Delta_c

        G31 = gamma31 - 1j * Delta31
        G41 = gamma41 - 1j * Delta41
        G21 = gamma21 - 1j * delta_two_photon

        # Correct sign convention:
        # With Omega_c2 = 0, this gives the usual denominator
        # G31 + |Omega_c1|^2 / (4 G21).
        A = np.array([
            [G31, 0.0, -1j * Omega_c1 / 2.0],
            [0.0, G41, -1j * Omega_c2 / 2.0],
            [-1j * np.conj(Omega_c1) / 2.0,
             -1j * np.conj(Omega_c2) / 2.0,
             G21]
        ], dtype=complex)

        b = np.array([
            1j * Omega_p1 / 2.0,
            1j * Omega_p2 / 2.0,
            0.0
        ], dtype=complex)

        rho31, rho41, rho21 = np.linalg.solve(A, b)

        chi_p = mu31 * rho31 + mu41 * rho41

        rho31_list.append(rho31)
        rho41_list.append(rho41)
        rho21_list.append(rho21)
        chi_list.append(chi_p)

    rho31_arr = np.array(rho31_list)
    rho41_arr = np.array(rho41_list)
    rho21_arr = np.array(rho21_list)
    chi_arr = np.array(chi_list)

    return {
        "rho31": rho31_arr,
        "rho41": rho41_arr,
        "rho21": rho21_arr,
        "chi": chi_arr,
        "absorption": np.imag(chi_arr),
        "dispersion": np.real(chi_arr),
        "Omega_p2": Omega_p2,
        "Omega_c2": Omega_c2,
    }


def absorption_contrast_at_center(
    Delta_p,
    absorption,
    absorption_ref_center
):
    """
    吸収ベースのEIT contrast:
        C = 1 - A_EIT(0) / A_no_control(0)

    C = 1 means perfect transparency.
    C = 0 means no EIT.
    """

    idx0 = np.argmin(np.abs(Delta_p))
    A0 = absorption[idx0]

    return 1.0 - A0 / absorption_ref_center


def dark_state_mismatch(r_p, r_c):
    """
    4準位でのdark-state破れの指標。
    r_p == r_c なら 0。
    """

    numerator = abs(r_p - r_c)
    denominator = np.sqrt((1 + abs(r_p)**2) * (1 + abs(r_c)**2))
    return numerator / denominator


if __name__ == "__main__":

    MHz = 1.0

    Delta_p = np.linspace(-1200, 1200, 5001) * MHz

    gamma31 = 1000 * MHz
    gamma41 = 1000 * MHz
    gamma21 = 1 * MHz

    Delta_e = 500 * MHz

    # weak probe: fixed
    Omega_p1 = 1.0 * MHz

    # leakage ratios
    # r_p: probe coupling ratio to |4>
    # r_c: control coupling ratio to |4>
    #
    # r_p != r_c gives incomplete dark-state cancellation.
    r_p = 0.08
    r_c = 0.50

    print("dark-state mismatch =", dark_state_mismatch(r_p, r_c))

    # no-control reference absorption
    ref = solve_eit_4level_leakage(
        Delta_p_array=np.array([0.0]),
        Omega_p1=Omega_p1,
        Omega_c1=0.0,
        r_p=r_p,
        r_c=0.0,
        gamma31=gamma31,
        gamma41=gamma41,
        gamma21=gamma21,
        Delta_e=Delta_e,
    )

    A_ref_center = ref["absorption"][0]

    # Compare Omega_c1 < Delta_e and Omega_c1 > Delta_e
    cases = [
        ("ideal 3-level, Omega_c=600", 600 * MHz, 0.0, 0.0),
        ("leakage, Omega_c=300 < Delta_e", 300 * MHz, r_p, r_c),
        ("leakage, Omega_c=600 > Delta_e", 600 * MHz, r_p, r_c),
        ("leakage, Omega_c=1000 > Delta_e", 1000 * MHz, r_p, r_c),
    ]

    plt.figure(figsize=(7, 4))

    for label, Omega_c1, rp_case, rc_case in cases:
        out = solve_eit_4level_leakage(
            Delta_p_array=Delta_p,
            Omega_p1=Omega_p1,
            Omega_c1=Omega_c1,
            r_p=rp_case,
            r_c=rc_case,
            gamma31=gamma31,
            gamma41=gamma41,
            gamma21=gamma21,
            Delta_e=Delta_e,
        )

        A = out["absorption"]

        # normalize by no-control line-center absorption
        A_norm = A / A_ref_center

        C = absorption_contrast_at_center(
            Delta_p=Delta_p,
            absorption=A,
            absorption_ref_center=A_ref_center
        )

        plt.plot(
            Delta_p / MHz,
            A_norm,
            label=f"{label}, C={C:.3f}"
        )

    plt.axvline(0, ls="--", alpha=0.4)
    plt.axvline(Delta_e / MHz, ls=":", alpha=0.4, label="extra level position")
    plt.xlabel("Probe detuning Δp (MHz)")
    plt.ylabel("Normalized absorption")
    plt.title("EIT contrast reduction by nearby excited state")
    plt.legend()
    plt.tight_layout()
    plt.show()

    # Sweep Omega_c1 / Delta_e
    Omega_c1_array = np.linspace(20, 2000, 120) * MHz

    contrast_ideal = []
    contrast_leak = []

    for Omega_c1 in Omega_c1_array:
        out_ideal = solve_eit_4level_leakage(
            Delta_p_array=Delta_p,
            Omega_p1=Omega_p1,
            Omega_c1=Omega_c1,
            r_p=0.0,
            r_c=0.0,
            gamma31=gamma31,
            gamma41=gamma41,
            gamma21=gamma21,
            Delta_e=Delta_e,
        )

        C_ideal = absorption_contrast_at_center(
            Delta_p,
            out_ideal["absorption"],
            A_ref_center
        )
        contrast_ideal.append(C_ideal)

        out_leak = solve_eit_4level_leakage(
            Delta_p_array=Delta_p,
            Omega_p1=Omega_p1,
            Omega_c1=Omega_c1,
            r_p=r_p,
            r_c=r_c,
            gamma31=gamma31,
            gamma41=gamma41,
            gamma21=gamma21,
            Delta_e=Delta_e,
        )

        C_leak = absorption_contrast_at_center(
            Delta_p,
            out_leak["absorption"],
            A_ref_center
        )
        contrast_leak.append(C_leak)

    plt.figure(figsize=(7, 4))
    plt.plot(Omega_c1_array / Delta_e, contrast_ideal, label="ideal 3-level")
    plt.plot(Omega_c1_array / Delta_e, contrast_leak, label="with nearby level")
    plt.axvline(1.0, ls="--", alpha=0.5, label=r"$\Omega_{c1}/\Delta_e=1$")
    plt.xlabel(r"$\Omega_{c1}/\Delta_e$")
    plt.ylabel("Absorption-based EIT contrast")
    plt.title("Contrast degradation when nearby excited state participates")
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import qutip
from qutip import basis, steadystate

def op(i, j, N=4):
    return basis(N, i) * basis(N, j).dag()

def eit_4level_full(
    Delta_p_array,
    Omega_p1,
    Omega_p2,
    Omega_c1,
    Omega_c2,
    Delta_e,
    Delta_c,
    Gamma31, Gamma32,
    Gamma41, Gamma42,
    gamma2phi=0.0,
    gamma3phi=0.0,
    gamma4phi=0.0,
    J34=0.0,
    mu31=1.0,
    mu41=1.0,
):
    """
    4準位EITのフル密度行列計算
    状態番号:
      |1> -> 0
      |2> -> 1
      |3> -> 2
      |4> -> 3
    """

    # projectors
    P1 = op(0, 0)
    P2 = op(1, 1)
    P3 = op(2, 2)
    P4 = op(3, 3)

    # transition operators
    s31 = op(2, 0)
    s41 = op(3, 0)
    s32 = op(2, 1)
    s42 = op(3, 1)
    s34 = op(2, 3)
    s43 = op(3, 2)

    # spontaneous decay operators
    c_ops_base = [
        np.sqrt(Gamma31) * op(0, 2),  # |3> -> |1>
        np.sqrt(Gamma32) * op(1, 2),  # |3> -> |2>
        np.sqrt(Gamma41) * op(0, 3),  # |4> -> |1>
        np.sqrt(Gamma42) * op(1, 3),  # |4> -> |2>
    ]

    # pure dephasing
    # ここでは coherence decay がだいたい gamma_phi になるように sqrt(2*gamma_phi) を使う
    # あなたの定義が異なるなら 2 を外してください
    if gamma2phi > 0:
        c_ops_base.append(np.sqrt(2 * gamma2phi) * P2)
    if gamma3phi > 0:
        c_ops_base.append(np.sqrt(2 * gamma3phi) * P3)
    if gamma4phi > 0:
        c_ops_base.append(np.sqrt(2 * gamma4phi) * P4)

    rho31_list = []
    rho41_list = []
    pop3_list = []
    pop4_list = []
    chi_list = []

    for Delta_p in Delta_p_array:
        delta = Delta_p - Delta_c

        H = (
            -delta * P2
            -Delta_p * P3
            -(Delta_p - Delta_e) * P4
            + 0.5 * (
                Omega_p1 * s31 + np.conj(Omega_p1) * s31.dag()
                + Omega_p2 * s41 + np.conj(Omega_p2) * s41.dag()
                + Omega_c1 * s32 + np.conj(Omega_c1) * s32.dag()
                + Omega_c2 * s42 + np.conj(Omega_c2) * s42.dag()
            )
            + J34 * (s34 + s43)   # 必要なら |3>-|4> 混合も追加
        )

        rho_ss = steadystate(H, c_ops_base)

        rho31 = rho_ss[2, 0]
        rho41 = rho_ss[3, 0]

        chi = mu31 * rho31 + mu41 * rho41

        rho31_list.append(rho31)
        rho41_list.append(rho41)
        pop3_list.append(np.real(rho_ss[2, 2]))
        pop4_list.append(np.real(rho_ss[3, 3]))
        chi_list.append(chi)

    return {
        "rho31": np.array(rho31_list),
        "rho41": np.array(rho41_list),
        "pop3": np.array(pop3_list),
        "pop4": np.array(pop4_list),
        "chi": np.array(chi_list),
        "absorption": np.imag(np.array(chi_list)),
        "dispersion": np.real(np.array(chi_list)),
    }


# ===== 使用例 =====
if __name__ == "__main__":
    MHz = 1.0

    Delta_p = np.linspace(-1200, 1200, 4001) * MHz

    # control が excited-state splitting を超える領域
    Omega_c1 = 700 * MHz
    Omega_c2 = 400 * MHz

    # 弱プローブにしておく
    Omega_p1 = 0.5 * MHz
    Omega_p2 = 0.2 * MHz

    Delta_e = 500 * MHz
    Delta_c = 0.0 * MHz

    out = eit_4level_full(
        Delta_p_array=Delta_p,
        Omega_p1=Omega_p1,
        Omega_p2=Omega_p2,
        Omega_c1=Omega_c1,
        Omega_c2=Omega_c2,
        Delta_e=Delta_e,
        Delta_c=Delta_c,
        Gamma31=3.0 * MHz,
        Gamma32=2.0 * MHz,
        Gamma41=3.0 * MHz,
        Gamma42=2.0 * MHz,
        gamma2phi=0.05 * MHz,
        gamma3phi=0.0 * MHz,
        gamma4phi=0.0 * MHz,
        J34=0.0,
        mu31=1.0,
        mu41=1.0,
    )

    abs_sig = out["absorption"]
    abs_norm = abs_sig / np.max(np.abs(abs_sig))

    plt.figure(figsize=(7, 4))
    plt.plot(Delta_p / MHz, abs_norm, label="total Im[chi]")
    plt.axvline(0, ls="--", alpha=0.5)
    plt.xlabel("Probe detuning (MHz)")
    plt.ylabel("Normalized absorption")
    plt.legend()
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(7, 4))
    plt.plot(Delta_p / MHz, np.imag(out["rho31"]), label="Im[rho31]  via |3>")
    plt.plot(Delta_p / MHz, np.imag(out["rho41"]), label="Im[rho41]  via |4>")
    plt.axvline(0, ls="--", alpha=0.5)
    plt.xlabel("Probe detuning (MHz)")
    plt.ylabel("Component signals")
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# =========================================================
# 1) 補助関数
# =========================================================
def smooth1d(y, n=7):
    """簡単な移動平均"""
    y = np.asarray(y, dtype=float)
    if n <= 1:
        return y.copy()
    kernel = np.ones(n) / n
    return np.convolve(y, kernel, mode="same")


def find_extremum_in_window(x, y, xmin, xmax, mode="max"):
    """
    区間 [xmin, xmax] 内の局所極値を探す
    mode = "max" or "min"
    見つからない場合は単純な最大/最小を返す
    """
    mask = (x >= xmin) & (x <= xmax)
    xx = np.asarray(x)[mask]
    yy = np.asarray(y)[mask]

    if len(xx) < 3:
        idx = np.argmax(yy) if mode == "max" else np.argmin(yy)
        return xx[idx], yy[idx], idx

    if mode == "max":
        cand = np.where((yy[1:-1] > yy[:-2]) & (yy[1:-1] >= yy[2:]))[0] + 1
        if len(cand) == 0:
            idx = np.argmax(yy)
        else:
            idx = cand[np.argmax(yy[cand])]
    elif mode == "min":
        cand = np.where((yy[1:-1] < yy[:-2]) & (yy[1:-1] <= yy[2:]))[0] + 1
        if len(cand) == 0:
            idx = np.argmin(yy)
        else:
            idx = cand[np.argmin(yy[cand])]
    else:
        raise ValueError("mode must be 'max' or 'min'")

    return xx[idx], yy[idx], idx


def measure_eit_contrast(
    detuning,
    absorption,
    center_guess=0.0,
    peak_window=None,
    center_window=None,
    exclude_center=10.0,
    smooth_pts=7,
):
    """
    detuning, absorption から EIT contrast を測る
    absorption は「大きいほど吸収が強い」符号で与えること
    """

    x = np.asarray(detuning, dtype=float)
    y = smooth1d(np.asarray(absorption, dtype=float), smooth_pts)

    if peak_window is None:
        peak_window = 0.35 * (x.max() - x.min())
    if center_window is None:
        center_window = max(20.0, 0.15 * peak_window)

    # 中央ディップ
    x_dip, A_dip, _ = find_extremum_in_window(
        x, y,
        center_guess - center_window,
        center_guess + center_window,
        mode="min"
    )

    # 左右のピーク
    xL, A_left, _ = find_extremum_in_window(
        x, y,
        center_guess - peak_window,
        x_dip - exclude_center,
        mode="max"
    )
    xR, A_right, _ = find_extremum_in_window(
        x, y,
        x_dip + exclude_center,
        center_guess + peak_window,
        mode="max"
    )

    A_side = 0.5 * (A_left + A_right)
    contrast = (A_side - A_dip) / max(A_side, 1e-15)

    return {
        "contrast": contrast,
        "x_dip": x_dip,
        "A_dip": A_dip,
        "x_left": xL,
        "A_left": A_left,
        "x_right": xR,
        "A_right": A_right,
        "A_side": A_side,
        "smoothed_absorption": y,
    }


# =========================================================
# 2) 4準位と3準位参照を比較しながらコントラストを走査
# =========================================================
def scan_contrast_vs_control(
    Delta_p_array,
    Omega_c1_values,
    Omega_p1,
    rc,          # rc = Omega_c2 / Omega_c1
    rp,          # rp = Omega_p2 / Omega_p1
    Delta_e,
    Delta_c,
    Gamma31, Gamma32,
    Gamma41, Gamma42,
    gamma2phi=0.0,
    gamma3phi=0.0,
    gamma4phi=0.0,
    J34=0.0,
    mu31=1.0,
    mu41=1.0,
    peak_window=None,
    center_window=None,
):
    """
    不要準位あり(4準位) と 不要準位なし(3準位参照) を比較
    すべて MHz 単位で入れてよい
    """

    contrast_4 = []
    contrast_3 = []
    contrast_ratio = []
    pop4_at_dip = []
    dip_position_4 = []
    dip_position_3 = []

    for Omega_c1 in Omega_c1_values:
        Omega_c2 = rc * Omega_c1
        Omega_p2 = rp * Omega_p1

        # ----- 4準位 -----
        out4 = eit_4level_full(
            Delta_p_array=Delta_p_array,
            Omega_p1=Omega_p1,
            Omega_p2=Omega_p2,
            Omega_c1=Omega_c1,
            Omega_c2=Omega_c2,
            Delta_e=Delta_e,
            Delta_c=Delta_c,
            Gamma31=Gamma31,
            Gamma32=Gamma32,
            Gamma41=Gamma41,
            Gamma42=Gamma42,
            gamma2phi=gamma2phi,
            gamma3phi=gamma3phi,
            gamma4phi=gamma4phi,
            J34=J34,
            mu31=mu31,
            mu41=mu41,
        )

        # あなたのコードでは absorption = Im[chi] なので、
        # 吸収ピークを正にしたければ符号を反転
        alpha4 = -out4["absorption"]

        m4 = measure_eit_contrast(
            detuning=Delta_p_array,
            absorption=alpha4,
            center_guess=Delta_c,
            peak_window=peak_window,
            center_window=center_window,
            exclude_center=10.0,
            smooth_pts=7,
        )

        idx_dip4 = np.argmin(np.abs(Delta_p_array - m4["x_dip"]))

        # ----- 3準位参照（|4> を切る） -----
        out3 = eit_4level_full(
            Delta_p_array=Delta_p_array,
            Omega_p1=Omega_p1,
            Omega_p2=0.0,
            Omega_c1=Omega_c1,
            Omega_c2=0.0,
            Delta_e=Delta_e,   # 値は何でもよいがそのまま置く
            Delta_c=Delta_c,
            Gamma31=Gamma31,
            Gamma32=Gamma32,
            Gamma41=Gamma41,
            Gamma42=Gamma42,
            gamma2phi=gamma2phi,
            gamma3phi=gamma3phi,
            gamma4phi=gamma4phi,
            J34=0.0,
            mu31=mu31,
            mu41=0.0,          # |4> 側は観測に入れない
        )

        alpha3 = -out3["absorption"]

        m3 = measure_eit_contrast(
            detuning=Delta_p_array,
            absorption=alpha3,
            center_guess=Delta_c,
            peak_window=peak_window,
            center_window=center_window,
            exclude_center=10.0,
            smooth_pts=7,
        )

        c4 = m4["contrast"]
        c3 = m3["contrast"]

        contrast_4.append(c4)
        contrast_3.append(c3)
        contrast_ratio.append(c4 / max(c3, 1e-15))
        pop4_at_dip.append(out4["pop4"][idx_dip4])
        dip_position_4.append(m4["x_dip"])
        dip_position_3.append(m3["x_dip"])

    return {
        "Omega_c1_values": np.array(Omega_c1_values),
        "contrast_4": np.array(contrast_4),
        "contrast_3": np.array(contrast_3),
        "contrast_ratio": np.array(contrast_ratio),  # 1 に近ければ劣化が小さい
        "pop4_at_dip": np.array(pop4_at_dip),
        "dip_position_4": np.array(dip_position_4),
        "dip_position_3": np.array(dip_position_3),
    }


# =========================================================
# 3) スペクトル例を描く
# =========================================================
def plot_example_spectra(
    Delta_p_array,
    Omega_c1_list,
    Omega_p1,
    rc, rp,
    Delta_e,
    Delta_c,
    Gamma31, Gamma32,
    Gamma41, Gamma42,
    gamma2phi=0.0,
    gamma3phi=0.0,
    gamma4phi=0.0,
    J34=0.0,
    mu31=1.0,
    mu41=1.0,
):
    plt.figure(figsize=(7, 4))

    for Omega_c1 in Omega_c1_list:
        out = eit_4level_full(
            Delta_p_array=Delta_p_array,
            Omega_p1=Omega_p1,
            Omega_p2=rp * Omega_p1,
            Omega_c1=Omega_c1,
            Omega_c2=rc * Omega_c1,
            Delta_e=Delta_e,
            Delta_c=Delta_c,
            Gamma31=Gamma31,
            Gamma32=Gamma32,
            Gamma41=Gamma41,
            Gamma42=Gamma42,
            gamma2phi=gamma2phi,
            gamma3phi=gamma3phi,
            gamma4phi=gamma4phi,
            J34=J34,
            mu31=mu31,
            mu41=mu41,
        )

        alpha = -out["absorption"]
        alpha = alpha / max(np.max(np.abs(alpha)), 1e-15)

        plt.plot(Delta_p_array, alpha, label=fr"$\Omega_{{c1}}={Omega_c1:.0f}$ MHz")

    plt.axvline(Delta_c, ls="--", alpha=0.5)
    plt.xlabel("Probe detuning (MHz)")
    plt.ylabel("Normalized absorption")
    plt.legend()
    plt.tight_layout()
    plt.show()


# =========================================================
# 4) Delta_e と Omega_c1 の2次元ヒートマップ
# =========================================================
def contrast_heatmap(
    Delta_p_array,
    Omega_c1_values,
    Delta_e_values,
    Omega_p1,
    rc, rp,
    Delta_c,
    Gamma31, Gamma32,
    Gamma41, Gamma42,
    gamma2phi=0.0,
    gamma3phi=0.0,
    gamma4phi=0.0,
    J34=0.0,
    mu31=1.0,
    mu41=1.0,
    peak_window=None,
    center_window=None,
):
    Z = np.zeros((len(Delta_e_values), len(Omega_c1_values)))

    for i, Delta_e in enumerate(Delta_e_values):
        for j, Omega_c1 in enumerate(Omega_c1_values):
            out = eit_4level_full(
                Delta_p_array=Delta_p_array,
                Omega_p1=Omega_p1,
                Omega_p2=rp * Omega_p1,
                Omega_c1=Omega_c1,
                Omega_c2=rc * Omega_c1,
                Delta_e=Delta_e,
                Delta_c=Delta_c,
                Gamma31=Gamma31,
                Gamma32=Gamma32,
                Gamma41=Gamma41,
                Gamma42=Gamma42,
                gamma2phi=gamma2phi,
                gamma3phi=gamma3phi,
                gamma4phi=gamma4phi,
                J34=J34,
                mu31=mu31,
                mu41=mu41,
            )

            alpha = -out["absorption"]
            m = measure_eit_contrast(
                detuning=Delta_p_array,
                absorption=alpha,
                center_guess=Delta_c,
                peak_window=peak_window,
                center_window=center_window,
                exclude_center=10.0,
                smooth_pts=7,
            )

            Z[i, j] = m["contrast"]

    return Z


# =========================================================
# 5) 実行例
# =========================================================
if __name__ == "__main__":
    MHz = 1.0

    # detuning 軸
    Delta_p_array = np.linspace(-1200, 1200, 4001) * MHz

    # 固定パラメータ
    Omega_p1 = 0.5 * MHz
    Delta_c = 0.0 * MHz
    Delta_e = 500.0 * MHz

    Gamma31 = 3.0 * MHz
    Gamma32 = 2.0 * MHz
    Gamma41 = 3.0 * MHz
    Gamma42 = 2.0 * MHz

    gamma2phi = 0.05 * MHz
    gamma3phi = 0.0 * MHz
    gamma4phi = 0.0 * MHz
    J34 = 0.0

    # 不要準位への結合の強さ
    # rc = Omega_c2 / Omega_c1
    # rp = Omega_p2 / Omega_p1
    rc = 0.60
    rp = 0.40

    # --------------------------------------------
    # A. まずスペクトルの見た目
    # --------------------------------------------
    plot_example_spectra(
        Delta_p_array=Delta_p_array,
        Omega_c1_list=[100, 250, 400, 550, 700],
        Omega_p1=Omega_p1,
        rc=rc,
        rp=rp,
        Delta_e=Delta_e,
        Delta_c=Delta_c,
        Gamma31=Gamma31, Gamma32=Gamma32,
        Gamma41=Gamma41, Gamma42=Gamma42,
        gamma2phi=gamma2phi,
        gamma3phi=gamma3phi,
        gamma4phi=gamma4phi,
        J34=J34,
        mu31=1.0, mu41=1.0,
    )

    # --------------------------------------------
    # B. contrast vs control Rabi
    # --------------------------------------------
    Omega_c1_values = np.linspace(50, 900, 35) * MHz

    result = scan_contrast_vs_control(
        Delta_p_array=Delta_p_array,
        Omega_c1_values=Omega_c1_values,
        Omega_p1=Omega_p1,
        rc=rc,
        rp=rp,
        Delta_e=Delta_e,
        Delta_c=Delta_c,
        Gamma31=Gamma31, Gamma32=Gamma32,
        Gamma41=Gamma41, Gamma42=Gamma42,
        gamma2phi=gamma2phi,
        gamma3phi=gamma3phi,
        gamma4phi=gamma4phi,
        J34=J34,
        mu31=1.0,
        mu41=1.0,
        peak_window=900.0,
        center_window=120.0,
    )

    plt.figure(figsize=(7, 4))
    plt.plot(result["Omega_c1_values"], result["contrast_3"], "o-", label="ideal 3-level")
    plt.plot(result["Omega_c1_values"], result["contrast_4"], "s-", label="4-level with leakage")
    plt.xlabel(r"Control Rabi $\Omega_{c1}$ (MHz)")
    plt.ylabel("EIT contrast")
    plt.legend()
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(7, 4))
    plt.plot(result["Omega_c1_values"], result["contrast_ratio"], "o-")
    plt.axhline(1.0, ls="--", alpha=0.5)
    plt.xlabel(r"Control Rabi $\Omega_{c1}$ (MHz)")
    plt.ylabel(r"Contrast ratio $C_{4lvl}/C_{3lvl}$")
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(7, 4))
    plt.plot(result["Omega_c1_values"], result["pop4_at_dip"], "o-")
    plt.xlabel(r"Control Rabi $\Omega_{c1}$ (MHz)")
    plt.ylabel(r"Population in unwanted level $|4\rangle$ at EIT dip")
    plt.tight_layout()
    plt.show()

    # --------------------------------------------
    # C. heatmap: Delta_e と Omega_c1
    # --------------------------------------------
    Delta_e_values = np.linspace(150, 1000, 45) * MHz

    Z = contrast_heatmap(
        Delta_p_array=Delta_p_array,
        Omega_c1_values=Omega_c1_values,
        Delta_e_values=Delta_e_values,
        Omega_p1=Omega_p1,
        rc=rc,
        rp=rp,
        Delta_c=Delta_c,
        Gamma31=Gamma31, Gamma32=Gamma32,
        Gamma41=Gamma41, Gamma42=Gamma42,
        gamma2phi=gamma2phi,
        gamma3phi=gamma3phi,
        gamma4phi=gamma4phi,
        J34=J34,
        mu31=1.0,
        mu41=1.0,
        peak_window=900.0,
        center_window=120.0,
    )

    plt.figure(figsize=(7, 5))
    plt.imshow(
        Z,
        origin="lower",
        aspect="auto",
        extent=[
            Omega_c1_values[0], Omega_c1_values[-1],
            Delta_e_values[0], Delta_e_values[-1]
        ]
    )
    plt.colorbar(label="EIT contrast")
    plt.xlabel(r"Control Rabi $\Omega_{c1}$ (MHz)")
    plt.ylabel(r"Excited-state splitting $\Delta_e$ (MHz)")
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from qutip import basis, steadystate

def op(i, j, N=4):
    return basis(N, i) * basis(N, j).dag()

def eit_4level_full(
    Delta_p_array,
    Omega_p1,
    Omega_p2,
    Omega_c1,
    Omega_c2,
    Delta_e,
    Delta_c,
    Gamma31, Gamma32,
    Gamma41, Gamma42,
    gamma2phi=0.0,
    gamma3phi=0.0,
    gamma4phi=0.0,
    J34=0.0,
    mu31=1.0,
    mu41=1.0,
):
    """
    4準位EITのフル密度行列計算
    状態番号:
      |1> -> 0
      |2> -> 1
      |3> -> 2
      |4> -> 3
    """

    P1 = op(0, 0)
    P2 = op(1, 1)
    P3 = op(2, 2)
    P4 = op(3, 3)

    s31 = op(2, 0)
    s41 = op(3, 0)
    s32 = op(2, 1)
    s42 = op(3, 1)
    s34 = op(2, 3)
    s43 = op(3, 2)

    c_ops_base = [
        np.sqrt(Gamma31) * op(0, 2),  # |3> -> |1>
        np.sqrt(Gamma32) * op(1, 2),  # |3> -> |2>
        np.sqrt(Gamma41) * op(0, 3),  # |4> -> |1>
        np.sqrt(Gamma42) * op(1, 3),  # |4> -> |2>
    ]

    if gamma2phi > 0:
        c_ops_base.append(np.sqrt(2 * gamma2phi) * P2)
    if gamma3phi > 0:
        c_ops_base.append(np.sqrt(2 * gamma3phi) * P3)
    if gamma4phi > 0:
        c_ops_base.append(np.sqrt(2 * gamma4phi) * P4)

    rho31_list = []
    rho41_list = []
    pop3_list = []
    pop4_list = []
    chi_list = []

    for Delta_p in Delta_p_array:
        delta = Delta_p - Delta_c

        H = (
            -delta * P2
            -Delta_p * P3
            -(Delta_p - Delta_e) * P4
            + 0.5 * (
                Omega_p1 * s31 + np.conj(Omega_p1) * s31.dag()
                + Omega_p2 * s41 + np.conj(Omega_p2) * s41.dag()
                + Omega_c1 * s32 + np.conj(Omega_c1) * s32.dag()
                + Omega_c2 * s42 + np.conj(Omega_c2) * s42.dag()
            )
            + J34 * (s34 + s43)
        )

        rho_ss = steadystate(H, c_ops_base)

        rho31 = complex(rho_ss[2, 0])
        rho41 = complex(rho_ss[3, 0])

        chi = mu31 * rho31 + mu41 * rho41

        rho31_list.append(rho31)
        rho41_list.append(rho41)
        pop3_list.append(np.real(rho_ss[2, 2]))
        pop4_list.append(np.real(rho_ss[3, 3]))
        chi_list.append(chi)

    rho31_arr = np.array(rho31_list)
    rho41_arr = np.array(rho41_list)
    chi_arr = np.array(chi_list)

    return {
        "rho31": rho31_arr,
        "rho41": rho41_arr,
        "pop3": np.array(pop3_list),
        "pop4": np.array(pop4_list),
        "chi": chi_arr,
        "absorption": np.imag(chi_arr),
        "dispersion": np.real(chi_arr),
    }


def normalize(y):
    ymax = np.max(np.abs(y))
    if ymax == 0:
        return y
    return y / ymax


def leakage_ratio(im_rho31, im_rho41, eps=1e-15):
    num = np.abs(im_rho41)
    den = np.abs(im_rho31) + np.abs(im_rho41) + eps
    return num / den


if __name__ == "__main__":
    MHz = 1.0

    Delta_p = np.linspace(-1500, 1500, 5001) * MHz

    # =========================
    # パラメータ
    # =========================
    Delta_e = 500 * MHz
    Delta_c = 0.0 * MHz

    # 高温相当として excited-state linewidth を大きめに
    Gamma31 = 1250.0 * MHz
    Gamma32 = 1250.0 * MHz
    Gamma41 = 1250.0 * MHz
    Gamma42 = 1250.0 * MHz

    gamma2phi = 1.0 * MHz
    gamma3phi = 0.0 * MHz
    gamma4phi = 0.0 * MHz

    # プローブは弱く固定した方が判断しやすい
    Omega_p1 = 0.5 * MHz
    Omega_p2 = 0.05 * MHz

    # control on
    Omega_c1_on = 700.0 * MHz
    Omega_c2_on = 10.0 * MHz

    # control off
    Omega_c1_off = 0.0 * MHz
    Omega_c2_off = 0.0 * MHz

    # =========================
    # control ON
    # =========================
    out_on = eit_4level_full(
        Delta_p_array=Delta_p,
        Omega_p1=Omega_p1,
        Omega_p2=Omega_p2,
        Omega_c1=Omega_c1_on,
        Omega_c2=Omega_c2_on,
        Delta_e=Delta_e,
        Delta_c=Delta_c,
        Gamma31=Gamma31,
        Gamma32=Gamma32,
        Gamma41=Gamma41,
        Gamma42=Gamma42,
        gamma2phi=gamma2phi,
        gamma3phi=gamma3phi,
        gamma4phi=gamma4phi,
        J34=0.0,
        mu31=1.0,
        mu41=1.0,
    )

    # =========================
    # control OFF
    # =========================
    out_off = eit_4level_full(
        Delta_p_array=Delta_p,
        Omega_p1=Omega_p1,
        Omega_p2=Omega_p2,
        Omega_c1=Omega_c1_off,
        Omega_c2=Omega_c2_off,
        Delta_e=Delta_e,
        Delta_c=Delta_c,
        Gamma31=Gamma31,
        Gamma32=Gamma32,
        Gamma41=Gamma41,
        Gamma42=Gamma42,
        gamma2phi=gamma2phi,
        gamma3phi=gamma3phi,
        gamma4phi=gamma4phi,
        J34=0.0,
        mu31=1.0,
        mu41=1.0,
    )

    # 符号規約によっては absorption = -Im[chi] の方が見やすい場合あり
    abs_on = out_on["absorption"]
    abs_off = out_off["absorption"]

    # もし dip/peak が直感と逆なら下2行を使ってください
    # abs_on = -out_on["absorption"]
    # abs_off = -out_off["absorption"]

    abs_on_norm = normalize(abs_on)
    abs_off_norm = normalize(abs_off)

    im_rho31_on = np.imag(out_on["rho31"])
    im_rho41_on = np.imag(out_on["rho41"])

    leak = leakage_ratio(im_rho31_on, im_rho41_on)

    # =========================
    # Plot 1: total spectrum (ON/OFF comparison)
    # =========================
    plt.figure(figsize=(8, 4.8))
    plt.plot(Delta_p / MHz, abs_off_norm, label="control OFF", lw=2)
    plt.plot(Delta_p / MHz, abs_on_norm, label="control ON", lw=2)
    plt.axvline(0, ls="--", alpha=0.5)
    plt.axvline(Delta_e / MHz, ls=":", alpha=0.6, label="bare |4> resonance")
    plt.xlabel("Probe detuning (MHz)")
    plt.ylabel("Normalized absorption")
    plt.title("Total spectrum: control ON vs OFF")
    plt.legend()
    plt.tight_layout()
    plt.show()

    # =========================
    # Plot 2: separated components
    # =========================
    plt.figure(figsize=(8, 4.8))
    plt.plot(Delta_p / MHz, normalize(im_rho31_on), label="Im[rho31] via |3>", lw=2)
    plt.plot(Delta_p / MHz, normalize(im_rho41_on), label="Im[rho41] via |4>", lw=2)
    plt.axvline(0, ls="--", alpha=0.5)
    plt.axvline(Delta_e / MHz, ls=":", alpha=0.6, label="bare |4> resonance")
    plt.xlabel("Probe detuning (MHz)")
    plt.ylabel("Normalized component")
    plt.title("Separated absorption pathways")
    plt.legend()
    plt.tight_layout()
    plt.show()

    # =========================
    # Plot 3: leakage ratio
    # =========================
    plt.figure(figsize=(8, 4.8))
    plt.plot(Delta_p / MHz, leak, lw=2)
    plt.axvline(0, ls="--", alpha=0.5)
    plt.axvline(Delta_e / MHz, ls=":", alpha=0.6)
    plt.xlabel("Probe detuning (MHz)")
    plt.ylabel(r"|Im[rho41]| / (|Im[rho31]| + |Im[rho41]|)")
    plt.title("Leakage contribution from |4>")
    plt.tight_layout()
    plt.show()

    # =========================
    # Plot 4: excited-state populations
    # =========================
    plt.figure(figsize=(8, 4.8))
    plt.plot(Delta_p / MHz, out_on["pop3"], label="pop(|3>)", lw=2)
    plt.plot(Delta_p / MHz, out_on["pop4"], label="pop(|4>)", lw=2)
    plt.axvline(0, ls="--", alpha=0.5)
    plt.axvline(Delta_e / MHz, ls=":", alpha=0.6)
    plt.xlabel("Probe detuning (MHz)")
    plt.ylabel("Population")
    plt.title("Excited-state populations (control ON)")
    plt.legend()
    plt.tight_layout()
    plt.show()
```
