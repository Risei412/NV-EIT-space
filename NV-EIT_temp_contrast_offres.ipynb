$$

$\lvert g_1\rangle \leftrightarrow \lvert e_1\rangle$

\gamma_{\mathrm{off}}

$$
\alpha_{\mathrm{EIT}}
\propto
\frac{
\gamma_{\mathrm{eff}}

+

}
+
\alpha_{\mathrm{off}}
$$

であるため、

$$
\gamma_{\mathrm{eff}} \uparrow
\quad
\Rightarrow
\quad
\alpha_{\mathrm{EIT}} \uparrow
\quad
\Rightarrow
\quad

\quad
\Rightarrow
\quad

\quad

\quad
\text{EIT コントラストが低下}
$$

## EIT contrast の定義

EIT contrast は、EIT がない場合の二準位吸収を基準として、EIT 中心で吸収がどれだけ低下したかで定義する。

$$
C_{\mathrm{EIT}}
=
1-
\frac{
\alpha_{\mathrm{EIT}}
}{
\alpha_{2\mathrm{level}}
}.
$$

ここで、EIT 中心での吸収を

$$
\alpha_{\mathrm{EIT}}
\propto
\frac{
\gamma_{\mathrm{eff}}
}{
\gamma_{e_1 g_2}\gamma_{\mathrm{eff}}
+
|\Omega_{c1}|^2/4
}
$$

とし、二準位吸収を

$$
\alpha_{2\mathrm{level}}
\propto
\frac{1}{\gamma_{e_1 g_2}}
$$

とおくと、

$$
C_{\mathrm{EIT}}
=
1
-
\gamma_{e_1 g_2}
\frac{
\gamma_{\mathrm{eff}}
}{
\gamma_{e_1 g_2}\gamma_{\mathrm{eff}}
+
|\Omega_{c1}|^2/4
}.
$$

したがって、

$$
C_{\mathrm{EIT}}
=
\frac{
|\Omega_{c1}|^2/4
}{
\gamma_{e_1 g_2}\gamma_{\mathrm{eff}}
+
|\Omega_{c1}|^2/4
}.
$$

ここで、有効緩和率は

$$
\gamma_{\mathrm{eff}}
=
\gamma_{g_1 g_2}
+
\frac{|\Omega_{c2}|^2}{4}
\frac{
\gamma_{\mathrm{off}}
}{
\Delta_{\mathrm{off}}^2+\gamma_{\mathrm{off}}^2
}.
$$

したがって、off-resonant coupling が強くなると

$$
|\Omega_{c2}| \uparrow
\quad \Rightarrow \quad
\gamma_{\mathrm{eff}} \uparrow
\quad \Rightarrow \quad
C_{\mathrm{EIT}} \downarrow .
$$

$$

を用いる。低温側では、添付コードに合わせて `gamma_ph_lowT = 10e6` を入れている。

ここで使う値は、Supplementary Fig. S3 に合わせて

$$
\Omega_{c1}/(2\pi)=7.8,\ 20,\ 100\ \mathrm{MHz}
$$

とし、隣接励起準位との detuning は低温での典型的な励起状態間隔

$$
\Delta_{\mathrm{off}}/(2\pi)\simeq 0.5\ \mathrm{GHz}
$$

を用いる。コード中では、全て `Hz` 単位、すなわち $\Omega/(2\pi)$, $\Delta/(2\pi)$, $\gamma/(2\pi)$ 相当の値として統一する。

$\Omega_{c2}$ は現象論的に

$$
\Omega_{c2}
=
\eta\,\Omega_{c1}
\left(
\frac{\Omega_{c1}}{\Omega_{\mathrm{ref}}}
\right)^\beta
$$

と置く。まずは $\eta=1,\ \beta=0$ を初期値とし、実験点に合わせる場合は `eta_c2`, `beta_c2`, `gamma_g_eff` を調整する。


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python


# ============================================================
# 1. Phonon-induced transition rate
# ============================================================

des_perp = 6.0e3      # Hz/(V/m)
alpha_ph = 1.4        # Hz/K^5
omega_c_eV = 60e-3    # hbar*omega_c = 60 meV
```

```python
# Supplementary Fig. S3 の 9-level model では transverse strain = 15 GHz 付近を仮定
strain_perp_Hz = 15e9
delta_perp = np.array([
    strain_perp_Hz / des_perp / np.sqrt(2),
    strain_perp_Hz / des_perp / np.sqrt(2),
    0.0
])
```

```python


def I_phonon(omega_c, T, delta, n_grid=100):
    """
    Dimensionless phonon integral.
    omega_c: eV
    T      : K
    delta  : V/m vector
    """
    if T <= 0:
        return 0.0

    h = 4.135e-15      # eV s
    k_B = 8.617e-5     # eV/K

    x_c = omega_c / (k_B * T)
    x_delta = h * 2 * des_perp * np.sqrt(delta[0]**2 + delta[1]**2) / (k_B * T)

    x = np.linspace(x_c / 100, 0.9 * x_c, n_grid)
    dx = x[1] - x[0]

    numerator = x * (x + x_delta) * (x**2 + (x + x_delta)**2)
    denominator = (np.exp(x) - 1) * (1 - np.exp(-(x + x_delta)))

    return np.sum(dx * numerator / denominator)

def gamma_phonon_scalar(T, delta=delta_perp, gamma_ph_lowT=10e6):
    """
    Phonon-induced rate in Hz.
    User-provided rule:
      T > 10 K: alpha_ph * I * T^5
      T <= 10 K: gamma_ph_lowT
    """
    if T > 10:
        return alpha_ph * I_phonon(omega_c_eV, T, delta) * T**5
    return gamma_ph_lowT

def gamma_phonon(T, delta=delta_perp, gamma_ph_lowT=10e6):
    T = np.asarray(T, dtype=float)
    return np.array([gamma_phonon_scalar(float(t), delta, gamma_ph_lowT) for t in T])


# ============================================================
# 2. EIT contrast with off-resonant coupling
# ============================================================

def gamma_eff_off(gamma_g, Omega_c2, Delta_off, gamma_off):
    """
    gamma_eff = gamma_g + off-resonant induced loss.
    All quantities are in Hz-like units, i.e. Omega/(2pi), Delta/(2pi), gamma/(2pi).
    """
    return gamma_g + (Omega_c2**2 / 4.0) * gamma_off / (Delta_off**2 + gamma_off**2)

def contrast_EIT_T(gamma_eg, gamma_g, Omega_c1, Omega_c2, Delta_off, gamma_off):
    geff = gamma_eff_off(gamma_g, Omega_c2, Delta_off, gamma_off)
    return (Omega_c1**2 / 4.0) / (gamma_eg * geff + Omega_c1**2 / 4.0)

def Omega_c2_phenomenology(Omega_c1, Omega_ref, eta=1.0, beta=0.0):
    """
    Phenomenological dependence of off-resonant coupling on resonant control Rabi frequency.

    eta  : relative off-resonant coupling strength at Omega_c1 = Omega_ref
    beta : nonlinearity. beta=0 gives Omega_c2 proportional to Omega_c1.
    """
    return eta * Omega_c1 * (Omega_c1 / Omega_ref)**beta


# ============================================================
# 3. Literature / fit parameters
# ============================================================

# Supplementary Fig. S3 values: Omega0/(2pi) = 7.8, 20, 100 MHz.
Omega_c1_list = [7.8e6, 20e6, 100e6]
Omega_ref = 7.8e6

# Low-temperature excited-state spacing in the supplement: Delta_e ≈ 0.5 GHz.
Delta_off_lit = 0.5e9

# Effective ground-state coherence decay used in this scalar model.
# The full 9-level paper fit gives gamma_gs/(2pi) ~ 240 kHz for mi=±1.
# In this reduced contrast-only formula, gamma_g_eff is treated as a phenomenological fit parameter.
gamma_g_eff = 1e5

# Phenomenological off-resonant coupling.
eta_c2 = 1.0
beta_c2 = 0.3

# Low-T phonon floor from the user-provided code.
gamma_ph_lowT = 10e6

# Optional local heating shift used in Supplementary Fig. S3(b).
# Set 0 for Fig. S3(a)-style cold-finger temperature.
local_temperature_shift_K = 7.0

# Overall reduction factor.
# The supplement reduces full 9-level simulated contrast by 13.
# For this scalar model, we instead normalize to the low-temperature experimental contrast.
target_lowT_contrast = 0.042
T_ref_for_normalization = 10.0


# ============================================================
# 4. Calculate curves
# ============================================================

T_cold = np.linspace(0, 40, 401)
T_local = T_cold + local_temperature_shift_K

gamma_eg_T = gamma_phonon(T_local, gamma_ph_lowT=gamma_ph_lowT)
gamma_off_T = gamma_eg_T.copy()

# Normalize the 7.8 MHz curve at T_ref to the observed low-T contrast.
gamma_ref = gamma_phonon(np.array([T_ref_for_normalization]), gamma_ph_lowT=gamma_ph_lowT)[0]
Omega_c1_ref = Omega_ref
Omega_c2_ref = Omega_c2_phenomenology(Omega_c1_ref, Omega_ref, eta_c2, beta_c2)
raw_ref = contrast_EIT_T(
    gamma_ref, gamma_g_eff,
    Omega_c1_ref, Omega_c2_ref,
    Delta_off_lit, gamma_ref
)
contrast_scale = target_lowT_contrast / raw_ref

curves = {}
for Omega_c1 in Omega_c1_list:
    Omega_c2 = Omega_c2_phenomenology(Omega_c1, Omega_ref, eta_c2, beta_c2)
    C_raw = contrast_EIT_T(
        gamma_eg_T, gamma_g_eff,
        Omega_c1, Omega_c2,
        Delta_off_lit, gamma_off_T
    )
    curves[Omega_c1] = contrast_scale * C_raw

print("Parameters used")
print(f"  Delta_off/(2pi)-like = {Delta_off_lit/1e9:.3f} GHz")
print(f"  gamma_g_eff/(2pi)-like = {gamma_g_eff/1e3:.1f} kHz")
print(f"  eta_c2 = {eta_c2:.3f}, beta_c2 = {beta_c2:.3f}")
print(f"  low-T contrast scale = {contrast_scale:.4f}")
print(f"  local temperature shift = {local_temperature_shift_K:.1f} K")


# ============================================================
# 5. Plot: Supplementary Fig. S3(b)-style x-axis
# ============================================================

plt.figure(figsize=(6.2, 4.4))

for Omega_c1, C in curves.items():
    plt.plot(
        T_local, C,
        label=fr"$\Omega_{{c1}}/(2\pi)={Omega_c1/1e6:g}\,\mathrm{{MHz}}$"
    )

plt.xlabel("Cold-finger temperature + 7 K (K)")
plt.ylabel("EIT contrast")
plt.title("Temperature-dependent EIT contrast with phonon broadening")
plt.ylim(0, 0.06)
plt.xlim(0, 47)
plt.grid(True, alpha=0.35)
plt.legend()
plt.show()


# ============================================================
# 6. Plot: cold-finger x-axis
# ============================================================

plt.figure(figsize=(6.2, 4.4))

for Omega_c1, C in curves.items():
    plt.plot(
        T_cold, C,
        label=fr"$\Omega_{{c1}}/(2\pi)={Omega_c1/1e6:g}\,\mathrm{{MHz}}$"
    )

plt.xlabel("Cold-finger temperature (K)")
plt.ylabel("EIT contrast")
plt.title("Temperature-dependent EIT contrast, plotted vs cold-finger temperature")
plt.ylim(0, 0.06)
plt.xlim(0, 40)
plt.grid(True, alpha=0.35)
plt.legend()
plt.show()
```


### パラメータ調整の読み方

このセルのモデルは Supplementary の 9-level density-matrix simulation そのものではなく、既存ノートブックの有効式に温度依存の phonon rate を入れた簡略モデルである。そのため、次のパラメータは現象論的な調整量として扱う。

- `gamma_g_eff`：基底状態コヒーレンスの有効緩和率。Supplementary の 3-level fit 値をそのまま使うなら `240e3` Hz だが、この一次元 contrast 式では減衰が強く出やすいため、まず `50e3` Hz 程度から調整する。
- `eta_c2`：$\Omega_{c2}/\Omega_{c1}$ の相対強度。大きくすると off-resonant coupling による contrast 低下が強くなる。
- `beta_c2`：$\Omega_{c2}$ の $\Omega_{c1}$ 依存性。`beta=0` なら線形、`beta>0` なら高 power 側で off-resonant coupling がより強くなる。
- `local_temperature_shift_K`：Supplementary Fig. S3(b) と同様に、ビームパスの局所温度が cold finger より高いと仮定する補正。
- `target_lowT_contrast`：実験で観測される低温側 contrast へ合わせるための全体スケール。Supplementary では full 9-level simulation に対して 13 分の 1 の縮小を用いているが、この簡略式では低温側の値へ直接規格化している。

実験点にさらに近づけるには、まず `gamma_g_eff` と `eta_c2` を変える。曲線全体が高温側に残りすぎる場合は `gamma_g_eff` または `eta_c2` を増やす。高 Rabi 周波数側だけが大きく残りすぎる場合は `beta_c2` を正にする。
