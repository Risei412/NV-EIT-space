NVモデルの記号で書くと、

$$

\Omega_{p1}

off-resonant control coupling は、主に有効二光子離調に AC Stark shift として入る：

$$
\Delta_{g_1 g_2}
=
\delta
+
i\gamma_{g_1 g_2}
-
\frac{|\Omega_{c1}|^2}{4\Delta_{e_1 g_2}}

ここで

$$
\delta = \Delta_p - \Delta_c
$$

$$
\Delta_{e_{\mathrm{off}} g_2}^{(c)}
$$

は control 光が $\lvert g_2\rangle \rightarrow \lvert e_{\mathrm{off}}\rangle$ に off-resonant に結合するときの離調である。

## NV-EIT 応答式の導出：off-resonant control coupling を含む場合

probe 光によって誘起される一次のコヒーレンスを

$$
\sigma_{e_1 g_2}^{(1)}
$$

とする。EIT 型の応答は、低次展開式ではなく、以下の非摂動的な分母構造を持つ形で扱う。

$$
\sigma_{e_1 g_2}^{(1)}
=
-
\frac{\rho_{g_2 g_2}^{(0)}}{2}
\frac{
\Delta_{g_1 g_2}^{\mathrm{NV}}
}{
\Delta_{e_1 g_2}
\Delta_{g_1 g_2}^{\mathrm{NV}}
-
|\Omega_{c1}|^2/4
}
\Omega_{p1}
-
\rho_{g_2 g_2}^{(0)}
N_{e_1 g_2}^{\mathrm{NV}}
\frac{\epsilon_p}{\hbar}.
$$

ここで、$\Omega_{p1}$ は probe 光による
$\lvert g_2\rangle \leftrightarrow \lvert e_1\rangle$
遷移の Rabi 周波数、$\Omega_{c1}$ は resonant control 光による
$\lvert g_1\rangle \leftrightarrow \lvert e_1\rangle$
遷移の Rabi 周波数である。

$\lvert e_{\mathrm{off}}\rangle$

$$
\Delta_{g_1 g_2}^{\mathrm{NV}}
=

+

-
\frac{|\Omega_{c2}|^2}

$$

とおく。

$$

$$

$$

$$

は基底状態間コヒーレンスの緩和率である。また、

$$
\Omega_{c2}
$$

は control 光が
$\lvert g_2\rangle \leftrightarrow \lvert e_{\mathrm{off}}\rangle$

$$
\Delta_{e_{\mathrm{off}}g_2}^{(c)}
$$

はその off-resonant 離調である。

$$

これが、隣接励起準位への off-resonant control coupling を含む NV-EIT の基本式である。

### 吸収スペクトル

probe 光に対する吸収は

$$
\mathrm{Im}

に比例する。したがって、

\alpha(\delta)

### EIT 中心での吸収

一光子共鳴付近で

$$
\Delta_{e_1 g_2}
=
i\gamma_{e_1 g_2}
$$

とし、EIT 中心を

$$
\delta_{\mathrm{eff}}=0
$$

とする。このとき

$$
\Delta_{g_1 g_2}^{\mathrm{NV}}
=
i\gamma_{\mathrm{eff}}
$$

であるため、

$$
\sigma_{e_1 g_2}^{(1)}
=
-
\frac{\rho_{g_2 g_2}^{(0)}}{2}
\frac{
i\gamma_{\mathrm{eff}}
}{
(i\gamma_{e_1 g_2})(i\gamma_{\mathrm{eff}})
-
|\Omega_{c1}|^2/4
}
\Omega_{p1}
-
\rho_{g_2 g_2}^{(0)}
N_{e_1 g_2}^{\mathrm{NV}}
\frac{\epsilon_p}{\hbar}.
$$

分母を整理すると、

$$
(i\gamma_{e_1 g_2})(i\gamma_{\mathrm{eff}})
-

=
-
\gamma_{e_1 g_2}\gamma_{\mathrm{eff}}
-

\sigma_{e_1 g_2}^{(1)}

### off-resonant coupling による EIT コントラスト低下

off-resonant coupling がない場合、

$$
\Omega_{c2}=0
$$

であり、

$$
\gamma_{\mathrm{eff}}
=
\gamma_{g_1 g_2}
$$

となる。

一方、off-resonant coupling がある場合、

$$
\gamma_{\mathrm{eff}}
=
\gamma_{g_1 g_2}
+
\frac{|\Omega_{c2}|^2}{4}
\frac{
\gamma_{\mathrm{off}}
}{
\Delta_{\mathrm{off}}^2+\gamma_{\mathrm{off}}^2
}
$$

である。したがって、$\Omega_{c2}$ が大きいほど、または
$|\Delta_{\mathrm{off}}|$ が小さいほど、

$$
\gamma_{\mathrm{eff}}
$$

は増大する。

EIT 中心の吸収は

$$
\alpha_{\mathrm{EIT}}
\propto
\frac{
\gamma_{\mathrm{eff}}
}{
\gamma_{e_1 g_2}\gamma_{\mathrm{eff}}
+
|\Omega_{c1}|^2/4
}
+
\alpha_{\mathrm{off}}
$$

であるため、

$$
\gamma_{\mathrm{eff}} \uparrow
\quad
\Rightarrow
\quad
\alpha_{\mathrm{EIT}} \uparrow
\quad
\Rightarrow
\quad
\text{EIT dip が浅くなる}.
$$

したがって、

$$
|\Delta_{\mathrm{off}}| \downarrow
\quad
\Rightarrow
\quad
\text{off-resonant coupling の影響が増大}
\quad
\Rightarrow
\quad
\text{EIT コントラストが低下}
$$

と解釈できる。

## EIT contrast の定義

EIT contrast は、EIT がない場合の二準位吸収を基準として、EIT 中心で吸収がどれだけ低下したかで定義する。

$$
C_{\mathrm{EIT}}
=
1-
\frac{
\alpha_{\mathrm{EIT}}
}{
\alpha_{2\mathrm{level}}
}.
$$

ここで、EIT 中心での吸収を

$$
\alpha_{\mathrm{EIT}}
\propto
\frac{
\gamma_{\mathrm{eff}}
}{
\gamma_{e_1 g_2}\gamma_{\mathrm{eff}}
+
|\Omega_{c1}|^2/4
}
$$

とし、二準位吸収を

$$
\alpha_{2\mathrm{level}}
\propto
\frac{1}{\gamma_{e_1 g_2}}
$$

とおくと、

$$
C_{\mathrm{EIT}}
=
1
-
\gamma_{e_1 g_2}
\frac{
\gamma_{\mathrm{eff}}
}{
\gamma_{e_1 g_2}\gamma_{\mathrm{eff}}
+
|\Omega_{c1}|^2/4
}.
$$

したがって、

$$
C_{\mathrm{EIT}}
=
\frac{
|\Omega_{c1}|^2/4
}{
\gamma_{e_1 g_2}\gamma_{\mathrm{eff}}
+
|\Omega_{c1}|^2/4
}.
$$

ここで、有効緩和率は

$$
\gamma_{\mathrm{eff}}
=
\gamma_{g_1 g_2}
+
\frac{|\Omega_{c2}|^2}{4}
\frac{
\gamma_{\mathrm{off}}
}{
\Delta_{\mathrm{off}}^2+\gamma_{\mathrm{off}}^2
}.
$$

したがって、off-resonant coupling が強くなると

$$
|\Omega_{c2}| \uparrow
\quad \Rightarrow \quad
\gamma_{\mathrm{eff}} \uparrow
\quad \Rightarrow \quad
C_{\mathrm{EIT}} \downarrow .
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gamma_eff(gamma_g, Omega_c2, Delta_off, gamma_off):
    return gamma_g + (Omega_c2**2 / 4) * gamma_off / (Delta_off**2 + gamma_off**2)

def alpha_EIT_center(gamma_eg, gamma_g, Omega_c1, Omega_c2, Delta_off, gamma_off):
    geff = gamma_eff(gamma_g, Omega_c2, Delta_off, gamma_off)
    return geff / (gamma_eg * geff + Omega_c1**2 / 4)

def contrast_EIT(gamma_eg, gamma_g, Omega_c1, Omega_c2, Delta_off, gamma_off):
    geff = gamma_eff(gamma_g, Omega_c2, Delta_off, gamma_off)
    return (Omega_c1**2 / 4) / (gamma_eg * geff + Omega_c1**2 / 4)

# example parameters
gamma_eg = 15.0      # optical coherence decay rate
gamma_g  = 0.01      # ground-state coherence decay rate
Omega_c1 = 8.0       # resonant control Rabi frequency
gamma_off = 15.0     # linewidth of off-resonant excited state

Omega_c2_array = np.linspace(0, 80, 400)
Delta_off = 100.0

alpha_array = alpha_EIT_center(
    gamma_eg, gamma_g, Omega_c1,
    Omega_c2_array, Delta_off, gamma_off
)

contrast_array = contrast_EIT(
    gamma_eg, gamma_g, Omega_c1,
    Omega_c2_array, Delta_off, gamma_off
)

plt.figure()
plt.plot(Omega_c2_array, alpha_array)
plt.xlabel(r"$\Omega_{c2}$")
plt.ylabel(r"$\alpha_{\mathrm{EIT}}$ at EIT center")
plt.title("Increase of residual absorption by off-resonant coupling")
plt.grid()
plt.show()

plt.figure()
plt.plot(Omega_c2_array, contrast_array)
plt.xlabel(r"$\Omega_{c2}$")
plt.ylabel("EIT contrast")
plt.title("EIT contrast reduction by off-resonant coupling")
plt.grid()
plt.show()
```

```python
Delta_off_array = np.linspace(5, 300, 400)
Omega_c2 = 40.0

contrast_vs_detuning = contrast_EIT(
    gamma_eg, gamma_g, Omega_c1,
    Omega_c2, Delta_off_array, gamma_off
)

plt.figure()
plt.plot(Delta_off_array, contrast_vs_detuning)
plt.xlabel(r"$\Delta_{\mathrm{off}}$")
plt.ylabel("EIT contrast")
plt.title("EIT contrast vs off-resonant detuning")
plt.grid()
plt.show()
```
