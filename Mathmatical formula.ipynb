#### 電気分極P

電気分極Pは、単位体積当たりの双極子モーメントの期待値で与えられる. (1)

$\mathbf{P} = \frac{1}{V} \Sigma_{i} < \mathbf{p_i}>$  (1)

ここで、$V$ は分子や原子を含む領域の体積であり、$< \mathbf{p_i}>$ は双極子モーメントの期待値である.

混合系に見いだされる物理量の期待値は、密度行列とその物理量に対応する演算子とのトレース(2)によって与えられる。

$< \mathbf{p_i}> = \mathbf{Tr} [\rho \mathbf{p}_{i} ]$ (2)

式(2)を計算すると、
$< \mathbf{p_i}> = \mu_{31} \rho_{13} + \mu_{32} \rho_{23}  + \mathrm{C.c}. $

### 吸収率(EIT信号)

物体の吸収率は、Kramas-kroninの分散関係式より、電器感受率$\chi$の虚部  $\mathrm{Im} [ \chi ]$ で与えられる.

電気偏極$\mathbf{P}$ は、$\mathbf{P} := \varepsilon_{0} \chi E$ で定義されるので、

$\chi = \frac{1}{\varepsilon_{0} E} \mathbf{P} = \frac{\hbar \mu}{\varepsilon_{0} \Omega}  \mathbf{P} $

である。なお、途中式の展開で、$\Omega =\mathbf{\mu} \cdot \mathbf{E} / \hbar$ を利用している.

#### 電気双極子相互作用ハミルトニアン(回転波近似)

$H_{ \mathrm{int} } := \mathbf{\mu} \cdot \mathbf{E}= - \frac{ \hbar }{ 2 }  \left\{ \Omega_{\mathrm{p}}(t) \mathrm{e}^{ \mathrm{i} \Delta_{1} t } |3 \rangle \langle 1 |  +  \Omega_{\mathrm{c}}(t) \mathrm{e}^{ \mathrm{i} \Delta_{2} t } |3 \rangle \langle 2 | + \mathrm{H.c}. \right\}$


行列表示すると以下のようになる.

$ H_{ \mathrm{int} }  = - \frac{ \hbar }{ 2 }
\begin{bmatrix}
   0 & 0 & \Omega_{\mathrm{p}}^{*}(t) \mathrm{e}^{ - \mathrm{i} \Delta_{1} t } & 0 \\
   0 & 0 & \Omega_{\mathrm{c}}^{*}(t) \mathrm{e}^{ - \mathrm{i} \Delta_{2} t } & 0 \\
   \Omega_{\mathrm{p}}(t) \mathrm{e}^{ \mathrm{i} \Delta_{1} t } & \Omega_{\mathrm{c}}(t) \mathrm{e}^{ \mathrm{i} \Delta_{2} t } & 0 & 0 \\
   0 & 0 & 0 & 0 \\
\end{bmatrix}
$

※　$ H_{ \mathrm{int} } := \mathbf{\mu} \cdot \mathbf{E}　$ に回転波近似に相当するユニタリ行列を用いてユニタリ変換をしている
